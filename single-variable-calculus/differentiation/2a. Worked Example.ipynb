Compute the tangent line to the curve $y=x^3-x$ at the point $(2,6)$

```octave
x = [-2.3:0.01:2.3];
y = x.^3 - x;

plot(x,y,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

Using the *power rule* and the *sum rule*, the derivative of the function is:

$\frac{d}{dx}y=3x^2-1$

So, the slope of the tangent line at $(2,6)$ is $3x^2-1$ when $x=2$, which equals $3\times2^2-1=11$

We also need the intercept to fully define the line, which we can get by using the point-slope form, given the point $(x_1,y_1) = (2,6)$:

$y-y_1=m(x-x_1)$

$y=m(x-x_1)+y_1$

For the intercept, we want the $y$ value when $x=0$, so if we plug that and the other values we already know we get:

$y=m(x-x_1)+y_1$

$y=11(0-2)+6$

$y=-16$

So, the tangent line at $(2,6)$ is $11x-16$.

```octave
x = [-2.3:0.01:2.3];
y = x.^3 - x;
tangent = 11*x-16;

plot(x,y,x,tangent,'LineWidth',2);
set(gca, "ylim", [-10 10]);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```
