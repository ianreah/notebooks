Informally: a limit is the value that a function **approaches** as it **goes towards** a certain $x$ value. (There is a more formal definition of a limit, but we probably don't need to get into that!)

In many cases the limit is simply **the value of the function** at that $x$ value, but there are situations where the limit can be different to the value of the function and even cases where the limit doesn't exist.

Consider the function: $f(x)=\frac{x^2-9}{x+3}$

```octave
x = [-10:0.25:10];
y = (x.^2-9) ./ (x+3);

plot(x,y);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

Notice the hole in the line at $x=-3$? That's because the value of the function is undefined at that point...

$f(-3)=\frac{-3^2-9}{-3+3}=\frac{0}{0}=\text{undefined}$

It's usually shown with an unfilled circle at the point where the value is undefined...

```octave
x = [-10:0.25:10];
y = (x.^2-9) ./ (x+3);

plot(x,y,-3,-6,"ob", "markersize", 16);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

By looking at the graph you can see that the function **approaches** $-6$ as the $x$ value **goes towards** $-3$. In other words, it looks like the value at $x=-3$ is **going to be** $-6$ it just happens that when we get there the value is actually undefined.

So, despite the fact that $f(-3)=\text{undefined}$, we can say that the limit of the function as $x$ goes towards $-3$ is $-6$. It's written as, $\lim\limits_{x \to -3}f(x)=-6$

Later we'll look at some ways of calculating the limit rather than having to read it off a graph.

## Continuity

So far we've said that in many (but not all) cases the limit of a function for a particular $x$ value is simply the value of the function at that point.

In fact, the limit will be equal to the value of the function at any point if the function is **continuous at that point** and it will be different at points where the function is **not continuous**. So, let's talk a little bit about continuity...

Very loosely, you can say that a function is continuous if you can draw its graph without having to take your pen off the paper! Consider the graph below:

![](img/continuity.jpg)

This is clearly **not** a continuous function...

* The point at $x=-3$ is similar to our first example. The value of the function is undefined at that point (shown by the unfilled circle) but the function **approaches** a particular value as $x$ goes towards $-3$ from either direction. This is called **removable discontinuity**.
* The function is also **not continuous** at $x=0$ but this time it approaches a **different** value as it approaches $x=0$ from each direction, i.e., the limit from the left (going towards $x=0$ from values where $x<0$) is different to the limit from the right (going towards $x=0$ from values where $x>0$). This is called **jump discontinuity**. In this case, the left limit is equal to $0$ and is written as $\lim\limits_{x \to 0^-}=0$. The right limit, therefore, is $\lim\limits_{x \to 0^+}=3$. Notice that the value of the function at $x=0$ is **defined** to be $3$ (shown by the filled circle at $(0,3)$ where the point at $(0,0)$ is unfilled).
* $x=2$ is another example of **removable discontinuity**. The difference this time being that the function has a **defined** value for $x=2$: the filled circle at $(2,-1)$. It's just different to the **limit** at that point.
* At $x=6$, the limit is equal to negative infinity from each direction. This is called **infinite discontinuity**.

In the above examples where the left and right limits aren't equal we say that the limit does not exist. Similarly, where a function increases or decreases infinitely we also say the limit does not exist.

Notice that the function is continuous at $x=5$: the function approaches the same value from either direction and the value it approaches is equal to the value of the function at that point. Of course, this is also true for any other point in the function between the examples given above. This gives us some "rules" for defining continuity...

For a function to be continuous at a point $x_0$ then:

1. The value of $f(x_0)$ must be defined
1. The left-hand limit and the right-hand limit of the function approaching $x_0$, i.e., $\lim\limits_{x \to x_0^+} f(x) = \lim\limits_{x \to x_0^-} f(x)$, must be equal
1. The limits from 2 and the defined value from 1 must also be equal, i.e., $\lim\limits_{x \to x_0^+} f(x) = \lim\limits_{x \to x_0^-} f(x) = f(x_0)$

## Numerically solving the value of a limit

**Substitution Method:**

The first thing to try is to just bung in the $x$ value of the limit to the function and see if you can calculate it:

$\lim\limits_{x \to -1}\frac{x^3+4x^2-3}{x^2+5}=\frac{(-1)^3+4(-1)^2-3}{(-1)^2+5}=\frac{0}{6}=0$

**Factoring Method:**

Sometimes, of course, you run into problem when you try the substitution method. Like the first example:

$f(x)=\frac{x^2-9}{x+3}$

...which gave us $\frac{0}{0}$ by substition when trying to calculate the limit when $x \to -3$. One trick is to try factoring both the numerator and denominator and see if any factor can be cancelled out...

$\lim\limits_{x \to -3}\frac{x^2-9}{x+3}=\lim\limits_{x \to -3}\frac{(x+3)(x-3)}{x+3}$

The factor $(x+3)$ can be cancelled out leaving a simpler expression that can be evaluated using substitution:

$\lim\limits_{x \to -3}(x-3)=-3-3=-6$

(The same value we got for the limit by reading off the graph!)

**Conjugate Method:**

The conjugate of a binomial expression (an expression with two terms) is the same expression but with the opposite sign between the terms. With the conjugate method we multiply and divide by the conjugate (so we're not changing the value of the expression). It's useful if the expression contains radicals (roots) because the product of conjugates containing radicals does not contain any radicals...

Consider $(\sqrt{x}-5)$. It's conjugate is $(\sqrt{x}+5)$, and their product contains no radical expressions:

$(\sqrt{x}-5)(\sqrt{x}+5)=(\sqrt{x})(\sqrt{x})-5\sqrt{x}+5\sqrt{x}-25=x-25$

Here's how you can use that to evaluate a limit:

$\lim\limits_{x \to 5}\frac{\sqrt{x+11}-4}{x-5}$

Multiply & divide by the conjugate of the binomial term in the numerator...

$\lim\limits_{x \to 5}\frac{(\sqrt{x+11}-4)(\sqrt{x+11}+4)}{(x-5)(\sqrt{x+11}+4)}$

$\lim\limits_{x \to 5}\frac{(x+11)-16}{(x-5)(\sqrt{x+11}+4)}$

$\lim\limits_{x \to 5}\frac{x-5}{(x-5)(\sqrt{x+11}+4)}$

Cancel the $x-5$ factor...

$\lim\limits_{x \to 5}\frac{1}{\sqrt{x+11}+4}$

...and solve by substition...

$\lim\limits_{x \to 5}\frac{1}{\sqrt{x+11}+4}=\frac{1}{\sqrt{5+11}+4}=\frac{1}{8}$
