An inverse function is a function that, when applied to the result of a particular function, returns the original input of the first function. In other words: take the result, $y$, of a function $f(x)$ (i.e., $y=f(x)$, then $g$ is an inverse function if $g(y)=x$.

The notation used for an inverse function is usually, $f^{-1}$.

## Graphing an inverse function: in general

Suppose you have a pair of values, $(x,y)$, such that $y$ is the result of applying the function $f$ to the input value $x$:

$y=f(x)$

Then we know that applying the inverse function, $f^{-1}$ to $y$ will give us $x$. So $(y,x)$ then is the pair of values such that $x$ is the result of applying the inverse function, $f^{-1}$ to $y$.

You can see then that these value pairs are co-ordinates on the graph of the function $f$ and its inverse function $f^{-1}$ respectively. In general, if you have the graph of a function $f$ you can find the graph of $f^{−1}$ by exchanging the $x-$ and $y-$coordinates of all the points on the graph.

In other words, the graph of $f^{−1}$ is the reflection of the graph of $f$ across the line $y=x$.

## Graphing an inverse function: an example

Suppose we have an equation:

$y=\sqrt{x}$

It might be helpful to simplify this to:

$y^2=x$

Using function notation this would be:

$f(x)=\sqrt{x}$

and

$f^{-1}(y)=y^2$

```octave
x=[0:0.01:1.5];
y=sqrt(x);

# Plot the inverse by swapping the x & y co-ordinates to "reflect" the line around y=x
plot(x,y,'LineWidth',2,y,x,'LineWidth',2,x,x,'color','k');
set(gca, "xlim", [0 1.5]);
set(gca, "ylim", [0 1.5]);

legend ({'f(x)=sqrt(x)','f^{-1}(x)', 'y=x'}, 'location', 'southeast');
legend boxoff;
```

```octave
# Just to be sure, let's plot f(x)=x^2 directly to make sure we get the same as the green line above
x=[0:0.01:1.5];
y=x.^2;

plot(x,y,'LineWidth',2);
set(gca, "xlim", [0 1.5]);
set(gca, "ylim", [0 1.5]);
```

This also suggests that the slope of the inverse function at any point will be the inverse of the slope of the original function at the same point. We can prove this with implicit differentiation as well...

$y=f(x)$

$f^{-1}(y)=x$

$\frac{d}{dx}(f^{-1}(y))=\frac{d}{dx}x=1$

By implicit differentiation (and the chain rule, of course):

$\frac{d}{dx}(f^{-1}(y))\frac{dy}{dx}=1$

$\frac{d}{dx}(f^{-1}(y))=\frac{1}{\frac{dy}{dx}}$

So, if we know the derivative of the original function we can easily find the derivative of the inverse function.

## Example: Derivative of arctan(x)

$y=\arctan(x)=\tan^{-1}(x)$

We get the inverse function by taking the $\tan$ of both sides:

$\tan(y)=\tan(\tan^{-1}(x))=x$

Before we try to find the derivative of that, let's remind ourselves [what the tan function looks like](0c. Trigonometric functions and the unit circle.ipynb#tan-and-sec)...

```octave
x=[-pi/2:0.01:pi/2];
t=tan(x);

plot(x,t,'LineWidth',3);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-1.6 1.6]);
set(gca, "ylim", [-10 10]);
set(gca,'xtick',[-pi/2:pi/2:pi/2])
set(gca,'xticklabel',{'-pi/2','0', 'pi/2'})
set(gca, "box", "off");
```

Then we can see what the inverse function $\arctan$ looks like by reflecting the graph of $\tan$ across the line $y=x$...

```octave
x=[-pi/2:0.01:pi/2];
t=tan(x);

xx=[-10:10];

plot(x,t,'LineWidth',3,t,x,'LineWidth',3,xx,xx,'color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-10 10]);
set(gca, "ylim", [-10 10]);
set(gca, "box", "off");

legend ({'tan(x)','arctan(x)', 'y=x'}, 'location', 'southeast');
legend boxoff;
```

We've [previously shown that](2. Using the definition to derive some differentiation rules.ipynb#tan)...

$\frac{d}{dy}\tan y=\sec^2y$

So, using implicit differentiation, we can now take the derivative of both sides of our equation...

$\tan y=x$

$\frac{d}{dx}(\tan y)=\frac{d}{dx}x$

(Chain rule) $\frac{d}{dy}(\tan y)\frac{dy}{dx}=1$

...then solve for $\frac{dy}{dx}$

$\frac{dy}{dx}=\frac{1}{\frac{d}{dy}(\tan y)}$

$\frac{dy}{dx}=\frac{1}{\sec^2y}$

Now, $\sec=\frac{1}{cos}$, therefore:

$\frac{dy}{dx}=\cos^2y$

But don't forget, that if you want the derivative as a function of $x$ you have to plug in the original formula for $y=\arctan (x)$...

$\frac{dy}{dx}=\cos^2(\arctan(x))$

We can go further with a little bit of trigonometry and pythagoras. If $\tan y=x$, then this can be represented in a right-angled triangle using the formual $\tan y=\frac{opposite}{adjacent}$, where $opposite=x$ and $adjacent=1$. The hypotenuse of this triangle, using pythagoras, is...

$h=\sqrt{1+x^2}$

Therefore:

$\cos y=\frac{adjacent}{hypotenuse}=\frac{1}{\sqrt{1+x^2}}$

$\cos^2 y =\frac{1}{1+x^2}$

And finally,

$\frac{d}{dx}\arctan(x)=\frac{1}{1+x^2}$

## Example 2: Derivative of arcsin(x)

Now we're getting the hang of it, let's rattle through another example.

$y=\arcsin(x)$

$\sin y=x$

$\frac{d}{dx}(\sin y)=\frac{d}{dx}x$

$(\cos y)y'=1$

$y'=\frac{1}{\cos y}$

To find the derivative in terms of $x$ use the [identity](0. References.ipynb#Trigonometric-identities):

$\cos^2y+\sin^2y=1$

$\cos^2y=1-\sin^2y$

$\cos y=\pm\sqrt{1-\sin^2y}$

Just choosing the positive branch:

$y'=\frac{1}{\cos y}=\frac{1}{\sqrt{1-\sin^2y}}$

$\sin y=x$ so $\sin^2 y=x^2$

$y'=\frac{1}{\sqrt{1-x^2}}$

Notice that choice we made between a positive and negative square root when solving for $\cos y$. You need to consider the [**domain** and **range**](0d.%20Domain%20and%20Range.ipynb) of the functions to make sure the choices make sense in the context of the problem, and if it helps, draw a graph of the problem.

The **domain** and **range** of the $\arcsin$ is usually defined as:

$\arcsin:[-1,1]\to[-\frac{\pi}{2},\frac{\pi}{2}]$

```octave
x=[-pi/2:0.01:pi/2];
s=sin(x);

plot(x,s,'LineWidth',3,s,x,'LineWidth',3);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-1.6 1.6]);
set(gca, "ylim", [-1.6 1.6]);
set(gca,'xtick',[-pi/2,-1,0,1,pi/2])
set(gca,'xticklabel',{'-pi/2','-1','0','1','pi/2'})
set(gca,'ytick',[-pi/2,-1,0,1,pi/2])
set(gca,'yticklabel',{'-pi/2','-1','0','1','pi/2'})
set(gca, "box", "off");

legend ({'sin(x)','arcsin(x)'});
legend boxoff;
```

Clearly the gradient of the $\arcsin$ function is always positive over this range so that's why, when we were faced with a choice in the calculations of the derivative we chose the option that would lead to positive values.

```octave
y=[-pi/2:0.01:pi/2];
x=sin(y);
d=1./sqrt(1-x.^2);

plot(x,y,'LineWidth',3,x,d,'LineWidth',3);

set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "ylim", [-3 3]);
set(gca, "box", "off");

legend ({'y=arcsin(x)','y'''}, 'location', 'southeast');
legend boxoff;
```
