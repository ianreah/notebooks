Find two values for $\theta$ so that:

$\frac{d}{d\theta}(\cos^2(\theta^4))=0$

---

First let's just plot it out so we can see what we're dealing with...(you don't need to do this to solve the problem but I think it's useful to picture the function as well)...

```octave
theta=[-2:0.001:2];
y=cos(theta.^4).^2;

plot(theta,y,'LineWidth',2);

set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

It's easy to see that we're going to need the [chain rule](2. Using the definition to derive some differentiation rules.ipynb#Chain-rule) to be able to differentiate this, but if you look closely you'll see it's actually a chain of **3** functions. So how do we differentiate a chain of 3 functions? Well, we'll start by introducing new variables for each of the individual functions...

$y=\cos^2(\theta^4)=(\cos(\theta^4))^2$

The innermost function raises to the power of 4, then we take the cosine and, finally, the outermost function takes the square. Let's name them, $f$, $g$ and $h$, starting with the outermost function...

$f(x)=x^2$

$g(x)=\cos(x)$

$h(x)=x^4$

So,

$y=f(g(h(\theta)))$

Now, we know from the [chain rule](2. Using the definition to derive some differentiation rules.ipynb#Chain-rule) how to find the derivative of a chain of two functions, so let's temporarily introduce another function, $k$, so that the overall function can be written in terms of just two functions...

$k(x)=g(h(x))$

Then,

$y=f(k(\theta))$

And we know how to differentiate that:

$\frac{dy}{d\theta}=f'(k(\theta)).k'(\theta)$

Then if we differentiate $k$ we can substitute...

$k'(x)=g'(h(x)).h'(x)$

$\frac{dy}{d\theta}=f'(g(h(\theta))).g'(h(\theta)).h'(\theta)$

(You can probably see the pattern here that you can use to differentiate any number of chained functions!)

So, using the actual functions we get

$\frac{dy}{d\theta}=2\cos(\theta^4).(-\sin(\theta^4)).4\theta^3$

$=-8\theta^3.\cos(\theta^4).\sin(\theta^4)$

And for fun let's plot that:

```octave
theta=[-2:0.001:2];
dy=-8.*(theta.^3).*cos(theta.^4).*sin(theta.^4);

plot(theta,dy,'LineWidth',2);

set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

Again, you don't need to do this to get the solution, but if we plot them on the same plot we can use it to convince ourselves if the derivative make sense compared to the gradient of the original function - and we can get a rough idea of values for $\theta$ where the gradient is $0$, (i.e., where the derivative crosses the x-axis).

```octave
theta=[-2:0.001:2];
y=cos(theta.^4).^2;
dy=-8.*(theta.^3).*cos(theta.^4).*sin(theta.^4);

plotyy(theta,y,theta,dy);
set(gca, "xaxislocation", "zero");
```

Finally, to find some values of $\theta$ where the derivative is $0$ we need to solve the equation...

$-8\theta^3.\cos(\theta^4).\sin(\theta^4)=0$

Clearly, for the product of 3 terms to equal $0$ then at least one of the terms must be equal to $0$:

$-8\theta^3=0$ ...or... $\cos(\theta^4)=0$ ...or... $\sin(\theta^4)=0$

The first one is easy! If $-8\theta^3=0$ then $\theta=0$. (That's why the question asks us for **two** values - it would be too easy if we just left it there.)

Let's look at the second term...

$\cos(\theta^4)=0$

Well, of course, the cosine function is $0$ when the input is $...,-\frac{3\pi}{2},-\frac{\pi}{2},\frac{\pi}{2},\frac{3\pi}{2},...$

```octave
theta4=[-7:0.01:7];
y=cos(theta4);

plot(theta4,y,'LineWidth',2);

set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

So, in our case, $\cos(\theta^4))=0$ when $\theta^4$ is one of those values, e.g., $\theta^4=\frac{\pi}{2}$ or $\theta=(\frac{\pi}{2})^{\frac{1}{4}}$. So to answer the original question: two values for $\theta$ where the derivative of the original function is $0$ are $\theta=0$ and $\theta=(\frac{\pi}{2})^{\frac{1}{4}}$ (=~1.12). (And you can see that this matches what we saw when plotting the deriviative. It crosses the x-axis at both of those points.)

For completeness, let's look at the final term. Similarly, the sine function is $0$ at $...-2\pi,-\pi,0,\pi,2\pi,...$

```octave
theta4=[-7:0.01:7];
y=sin(theta4);

plot(theta4,y,'LineWidth',2);

set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

So, in our case, $\sin(\theta^4))=0$ when $\theta^4$ is one of those values.

If we put them together we can see if we get values that match what we're seeing in the plot of the derivative...

The derivative is $0$ when:

$\theta^4=0,\frac{\pi}{2},\pi,\frac{3\pi}{2},2\pi,...$

$\theta=0,(\frac{\pi}{2})^{\frac{1}{4}},\pi^{\frac{1}{4}},(\frac{3\pi}{2})^{\frac{1}{4}},2\pi^{\frac{1}{4}},...$

$\theta\approx0,1.12,1.33,1.47,1.58,...$

(And the same with negative values.)

Which does match the locations where a plot of the derivative crosses the x-axis...


```octave
theta=[0:0.001:1.6];
dy=-8.*(theta.^3).*cos(theta.^4).*sin(theta.^4);

plot(theta,dy,'LineWidth',2);

set(gca, "xlim", [0 1.6]);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```
