This is a bit trickier to derive so I'm going to work through step-by-step in a separate section of its own.

## Step 1

As you probably guessed the first step is to write out the derivative as the difference quotient of $a^x$.

$\frac{d}{dx}a^x=\lim\limits_{\Delta x\to0}\frac{a^{x+\Delta x}-a^x}{\Delta x}$

Using the [exponent property](0.%20References.ipynb#Exponents-refresher) $a^{x+\Delta x}=a^xa^{\Delta x}$

$\frac{d}{dx}a^x=\lim\limits_{\Delta x\to0}\frac{a^xa^{\Delta x}-a^x}{\Delta x}$

$=\lim\limits_{\Delta x\to0}a^x\frac{a^{\Delta x}-1}{\Delta x}$

Note that, in this limit $a^x$ is constant; it's $\Delta x$ that is changing. So, we can use the [constant multiple rule of limits](0.%20References.ipynb#Limits-rules)...

$\frac{d}{dx}a^x=a^x\lim\limits_{\Delta x\to0}\frac{a^{\Delta x}-1}{\Delta x}$

## Step 2

So, the derivative of $a^x$ is $a^x$ times by some multiple, dependant on $a$. Let's call that multiple $M(a)$...

$M(a)=\lim\limits_{\Delta x\to0}\frac{a^{\Delta x}-1}{\Delta x}$

$\frac{d}{dx}a^x=a^xM(a)$

If we plug in $x=0$...

$\frac{d}{dx}a^x|_{x=0}=a^0M(a)$

...and $a^0=1$, so...

$\frac{d}{dx}a^x|_{x=0}=M(a)$

In other words, $M(a)$ is the *slope* of the graph of $y=a^x$ at $x=0$. If we knew this we could then work out the slope at any point of the graph by multiplying it by $a^x$.

## Step 3

Define a number, $e$, such that $M(e)=1$. Therefore,

$\frac{d}{dx}e^x=e^xM(e)=e^x.1=e^x$

$\frac{d}{dx}e^x=e^x$

...and, to be consistent with our previous definition of $M$, the slope of $y=e^x$ at $x=0$ should be $1$. We can confirm this by plugging it straight in...

$\frac{d}{dx}e^x|_{x=0}=e^0=1$

We don't know what $e$ is yet, but before we can proceed we need to convince ourselves that such a number does exist.

Note that as the base, $a$, increases then the slope of the curve $y=a^x$ increases...

```octave
x=[0:0.01:1.5];
y1=1.^x;
y2=2.^x;
y3=3.^x;
y4=4.^x;
y5=5.^x;

plot(x,y1,x,y2,x,y3,x,y4,x,y5);
set(gca, "xlim", [-0 1.5]);
set(gca, "box", "off");
```

We can also make some very rough observations about some of the slopes at $x=0$. Consider the graph for $y=2^x$ (i.e., $a=2$)...

```octave
x=[-0.5:0.01:1.5];
y=2.^x;

plot(x,y,'LineWidth',3,x,x+1,'color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "ylim", [-0.5 3]);
set(gca, "box", "off");
```

A secant line with slope $1$ going through $(0,1)$ shows that the slope at $x=0$ of $y=2^x$ is **less than** $1$. Now let's look at $y=4^x$ (i.e., $a=4$)...

```octave
x=[-0.75:0.01:1];
y=4.^x;

plot(x,y,'LineWidth',3,x,x+1,'color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-0.75 1]);
set(gca, "ylim", [-0.5 3]);
set(gca, "box", "off");
```

This time the secant line shows that the slope is **greater than** $1$ at $x=0$.

So, if...

* the slope increases as $a$ increases, and,
* the slope is **less than** $1$ when $a=2$, and,
* the slope is **greater than** $1$ when $a=4$...

...then it's safe to assume that a number $e$ exists such that the slope of $y=e^x$ at $x=0$ is $1$, (and that this number is somewhere between $2$ and $4$).

## Step 4

Consider $y=e^x$ and it's inverse $w=\ln(x)$

```octave
x=[-1:0.01:4];
y=e.^x;

plot(x,y,'LineWidth',3,y,x,'LineWidth',3,x,x,'color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-1 4]);
set(gca, "ylim", [-1 4]);
set(gca, "box", "off");
```

Now use implicit differentiation to find the derivative of $\ln(x)$

$w=\ln(x)$

$e^w=x$

$\frac{d}{dx}e^w=1$

$e^ww'=1$ (because, $\frac{d}{dx}e^x=e^x$, by definition, remember?)

$w'=\frac{1}{e^w}$ (and, $e^w=x$, so...)

$w'=\frac{1}{x}$ 

## Step 5

Convert $a^x$ to base $e$...

$a=e^{\ln(a)}$

$a^x=\big(e^{\ln(a)}\big)^x$

$a^x=e^{x\ln(a)}$ (by [exponent's power of a power rule](0.%20References.ipynb#Exponents-refresher))

So we can differentiate that with the chain rule:

$\frac{d}{dx}a^x=\frac{d}{dx}e^{x\ln(a)}$

Suppose:

$g(x)=x\ln(a)$

$f(x)=e^x$

...then the function we're trying to differentiate is $f(g(x))$. (Remember $a$ is just a constant.)

$g'(x)=\ln(a)$

$f'(x)=e^x$

So,

$\frac{d}{dx}e^{x\ln(a)}=f'(g(x))g'(x)=e^{x\ln(a)}\ln(a)$

But, $e^{x\ln(a)}=a^x$, so...

$\frac{d}{dx}a^x=\frac{d}{dx}e^{x\ln(a)}=a^x\ln(a)$

Remember from Step 2 we said:

$\frac{d}{dx}a^x=a^xM(a)$

So now we know that $M(a)=\ln(a)$

$\frac{d}{dx}a^x=a^x\ln(a)$
