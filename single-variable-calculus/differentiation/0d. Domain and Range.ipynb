## $sin^{-1}$ ##

If you plot the inverse of the $\sin$ function by reflecting it around the line $y=x$, ([see later](3a.%20Implicit%20Differentiation%20%26%20Inverse%20Functions.ipynb)), then result is **not** a function...

```octave
x=[-2*pi:0.01:2*pi];
s=sin(x);

plot(x,s,'LineWidth',3,s,x,'LineWidth',3,x,x,'Color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.5 6.5]);
set(gca, "ylim", [-6.5 6.5]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0','pi/2','pi','3pi/2','2pi'})
set(gca,'ytick',[-2*pi,-3*pi/2,-pi,-pi/2:pi/2:pi/2,pi,3*pi/2,2*pi])
set(gca,'yticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0','pi/2','pi','3pi/2','2pi'})
set(gca, "box", "off");

legend ({'sin(x)','arcsin(x)', 'y=x'}, 'location', 'southeast');
legend boxoff;
```

You can tell it's not a function because for any input value, (between $−1$ and $1$), there will be multiple (I guess actually infinite!) points on the $\arcsin$ line as plotted above. A function is not allowed to have such ambiguity. An input value must give exactly one output.

To get round this we arbitrarily restrict the inputs and outputs to remove the ambiguity and so we can treat it as a function. The restriction of the inputs is called the **domain** and the set of possible output is called the **range**.

For the $\arcsin$ function we usually define the **domain** to be between from $−1$ to $1$ and the **range** to be from $-\frac{\pi}{2}$ to $\frac{\pi}{2}$. (That's the largest interval, centred around $0$, where there is no ambiguity in the output.)

```octave
x=[-pi/2:0.01:pi/2];
s=sin(x);

plot(x,s,'LineWidth',3,s,x,'LineWidth',3);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-1.6 1.6]);
set(gca, "ylim", [-1.6 1.6]);
set(gca,'xtick',[-pi/2,-1,0,1,pi/2])
set(gca,'xticklabel',{'-pi/2','-1','0','1','pi/2'})
set(gca,'ytick',[-pi/2,-1,0,1,pi/2])
set(gca,'yticklabel',{'-pi/2','-1','0','1','pi/2'})
set(gca, "box", "off");

legend ({'sin(x)','arcsin(x)'});
legend boxoff;
```

The notation we use to define the **domain** and **range** is...

$\arcsin:[-1,1]\to[-\frac{\pi}{2},\frac{\pi}{2}]$

## $cos^{-1}$ ##

We do a similar thing for the $\arccos$ function. You can see that the **domain** is still between $-1$ and $1$ but the **range** can't be centred around $0$ because that still results in ambiguous outputs...

```octave
x=[-2*pi:0.01:2*pi];
c=cos(x);

plot(x,c,'LineWidth',3,c,x,'LineWidth',3,x,x,'Color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.5 6.5]);
set(gca, "ylim", [-6.5 6.5]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0','pi/2','pi','3pi/2','2pi'})
set(gca,'ytick',[-2*pi,-3*pi/2,-pi,-pi/2:pi/2:pi/2,pi,3*pi/2,2*pi])
set(gca,'yticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0','pi/2','pi','3pi/2','2pi'})
set(gca, "box", "off");

legend ({'cos(x)','arccos(x)', 'y=x'}, 'location', 'southeast');
legend boxoff;
```

Instead we choose a **range** of $0$ to $\pi$ for the $\arccos$ function...

```octave
x=[0:0.01:pi];
c=cos(x);

plot(c,x,'LineWidth',3);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-1.2 1.2]);
set(gca, "ylim", [-0.2 3.3]);
set(gca,'ytick',[0,pi/2,pi])
set(gca,'yticklabel',{'0','pi/2','pi'})
set(gca, "box", "off");
```

$\arccos:[-1,1]\to[0,\pi]$

## $tan^{-1}$ ##

```octave
x=[-pi/2:0.01:pi/2];
t=tan(x);

xx=[-10:10];

plot(x,t,'LineWidth',3,t,x,'LineWidth',3,xx,xx,'color','k');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-10 10]);
set(gca, "ylim", [-10 10]);
set(gca, "box", "off");

legend ({'tan(x)','arctan(x)', 'y=x'}, 'location', 'southeast');
legend boxoff;
```

$\arctan:[-\infty,\infty]\to[-\frac{\pi}{2},\frac{\pi}{2}]$
