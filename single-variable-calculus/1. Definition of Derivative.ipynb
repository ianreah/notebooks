**In words:**

* The derivative of a function gives you the instantaneous rate of change of the function at any point
* Or, the derivative of a function gives you the slope of a tangent line at any point

**Mathematically:**

Think of the slope of a tangent line as the slope between two points on the function, where the distance between the two points approaches 0.

![](https://ocw.mit.edu/courses/mathematics/18-01-single-variable-calculus-fall-2006/18-01f06.jpg)

...and, remember that the slope of a line is $\frac{rise}{run}$

![](img\difference-quotient.jpg)

The derivative of a function is then defined as *the limit of a difference quotient*:

$$f'(x)=\lim\limits_{\Delta x\to 0} \frac{\Delta f}{\Delta x}=\lim\limits_{\Delta x\to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$$

-----


###Example

Calculate the derivative of the function:

$$f(x)=\frac{1}{1+x^2}$$

```octave
xs = [-5:0.01:5];
ys =1 ./ (1 +  xs.^2);
plot(xs,ys);
```

If we wanted to compute the derivative, $f'(x)$, so far all we've got is the limit definition of the derivative, i.e., the limit of a difference quotient:

$$f'(x)=\lim_{\Delta x\to 0} \frac{\Delta f}{\Delta x}=\lim_{\Delta x\to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$$

In this example:

$$f'(x)=\lim_{\Delta x\to 0} \frac{\frac{1}{1+(x+\Delta x)^2} - \frac{1}{1 + x^2}}{\Delta x}$$

But we can't just plug in $\Delta x = 0$, because it always resolves to $\frac {0}{0}$!

Pull out the $\frac{1}{\Delta x}$ and expand...

$$f'(x)=\lim_{\Delta x\to 0} \frac{1}{\Delta x} \Bigg(\frac{1}{1+(x+\Delta x)^2} - \frac{1}{1 + x^2}\Bigg)$$

$$=\lim_{\Delta x\to 0} \frac{1}{\Delta x} \Bigg(\frac{(1 + x^2)-(1+(x+\Delta x)^2)}{(1+(x+\Delta x)^2)(1 + x^2)}\Bigg)$$

$$=\lim_{\Delta x\to 0} \frac{1}{\Delta x} \Bigg(\frac{1 + x^2-1-x^2-2x\Delta x-\Delta x^2}{(1+(x+\Delta x)^2)(1 + x^2)}\Bigg)$$

$$=\lim_{\Delta x\to 0} \frac{1}{\Delta x} \Bigg(\frac{\Delta x(-2x-\Delta x)}{(1+(x+\Delta x)^2)(1 + x^2)}\Bigg)$$

...and the $\Delta x$ in the numerator cancels with the $\frac{1}{\Delta x}$...

$$f'(x)=\lim_{\Delta x\to 0} \Bigg(\frac{-2x-\Delta x}{(1+(x+\Delta x)^2)(1 + x^2)}\Bigg)$$

$$=\frac{-2x}{(1+x^2)^2}$$

$$\frac{d}{dx}\Bigg(\frac{1}{1+x^2}\Bigg)=\frac{-2x}{(1+x^2)^2}$$


```octave
xs = [-5:0.01:5];
ys = -(2 * xs) ./ ((1 +  xs.^2).^2);
plot(xs,ys);
```
