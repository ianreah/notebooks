## Power rule


$\frac{d}{dx}x^n=n.x^{n-1}$

**Proof:**

Let $f(x)=x^n$...

[By definition](1. Definition of Derivative.ipynb), the derivitave is equal to the limit of the difference quotient as $\Delta x \to 0$

$f'(x)=\lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)-f(x)}{\Delta x}$

$=\lim\limits_{\Delta x \to 0} \frac{(x + \Delta x)^n-x^n}{\Delta x}$

$=\lim\limits_{\Delta x \to 0} \frac{1}{\Delta x} ((x + \Delta x)^n-x^n)$

$(x + \Delta x)$ is a binomial (cos it has 2 terms), and the [binomial theorem](https://en.wikipedia.org/wiki/Binomial_theorem) gives us the pattern for the expansion of raising a binomial to a power, $n$. It's a little cumbersome to show the full expansion for a general power, $n$, but we actually only need the first 2 terms. So, we'll just show those to keep it simple...

$(x + \Delta x)^n=x^n + n.x^{n-1}.\Delta x + ...$

You'll see why we can get away with this later, but here's a clue: each term in the binomial expansion is of the form $ax^b\Delta x^c$ and $c$ will be a minimum of $2$ in each term we're omitting. (Actually, $2..n$ for the third and subsequent terms.)

$f'(x)=\lim\limits_{\Delta x \to 0} \frac{1}{\Delta x} (-x^n+x^n + n.x^{n-1}.\Delta x + ...)$

$=\lim\limits_{\Delta x \to 0} \frac{1}{\Delta x} (n.x^{n-1}.\Delta x + ...)$

$=\lim\limits_{\Delta x \to 0} (n.x^{n-1} + ...)$

Notice in that last line that we cancelled out the $\frac{1}{\Delta x}$ with the $\Delta x$ in the binomial term we're showing. Of course, we also have to do that with every term we're not showing. In fact, as we said above, each term we're not showing is multiplied by $\Delta x$ raised to *at least* the power of 2 so one of these will be cancelled from each term meaning that they will all still be multiplied by *at least* one $\Delta x$. So, when we take the limit, $\Delta x \to 0$, it means that the terms we're not showing are all $0$...

$f'(x)=\lim\limits_{\Delta x \to 0} (n.x^{n-1} + ...)$

$=n.x^{n-1} + 0$

$=n.x^{n-1}$

## Sum rule

$\frac{d}{dx}(f(x)+g(x))=f'(x)+g'(x)$

**Proof:**

Let, $h(x)=f(x)+g(x)$...

$h'(x)= \lim\limits_{\Delta x \to 0} \frac{h(x + \Delta x)-h(x)}{\Delta x}$

But, seeing as $h(x)=f(x)+g(x)$ we can expand it out to...

$h'(x)= \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)+g(x + \Delta x)-f(x)-g(x)}{\Delta x}$

$h'(x)= \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)-f(x)+g(x + \Delta x)-g(x)}{\Delta x}$

$h'(x)= \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)-f(x)}{\Delta x} + \frac{g(x + \Delta x)-g(x)}{\Delta x}$

The [limits sum rule](0. References.ipynb#Limits-rules) says that the limit of the sum of two expressions is equal the sum of their limits...

$h'(x)= \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)-f(x)}{\Delta x} + \lim\limits_{\Delta x \to 0} \frac{g(x + \Delta x)-g(x)}{\Delta x}$

Each limit above is the definition of $f'(x)$ and $g'(x)$ respectively. So,

$h'(x)= f'(x) + g'(x)$

## Constant multiple rule

$\frac{d}{dx}(c.f(x))=c.f'(x)$

**Proof:**

Let, $g(x)=c.f(x)$...

$g'(x)= \lim\limits_{\Delta x \to 0} \frac{g(x + \Delta x)-g(x)}{\Delta x}$

$g'(x)= \lim\limits_{\Delta x \to 0} \frac{c.f(x + \Delta x)-c.f(x)}{\Delta x}$

$g'(x)= \lim\limits_{\Delta x \to 0} c \frac{f(x + \Delta x)-f(x)}{\Delta x}$

The [limits constant multiple rule](0. References.ipynb#Limits-rules) says that the limit of a constant times an expressions is equal the constant times the limit of the expression...

$g'(x)= c \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x)-f(x)}{\Delta x}$

$g'(x)= c.f(x)$

## Trigonometry rules

### sin

$\frac{d}{dx}\sin x = \cos x$

**Proof:**

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \frac{\sin(x+\Delta x)-\sin(x)}{\Delta x}$

Using the [angle sum & difference trigonometric identities](0. References.ipynb#Trigonometric-identities)...

$\sin(a+b)=\sin(a)\cos(b)+\cos(a)\sin(b)$

$\sin(x+\Delta x)=\sin(x)\cos(\Delta x)+\cos(x)\sin(\Delta x)$

Therefore:

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \frac{\sin(x)\cos(\Delta x)+\cos(x)\sin(\Delta x)-\sin(x)}{\Delta x}$

With a little bit of shuffling around and factoring. (You'll see why shortly)...

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \Big(\frac{\sin(x)\cos(\Delta x)-\sin(x)}{\Delta x}+\frac{\cos(x)\sin(\Delta x)}{\Delta x}\Big)$

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \Big(\sin(x)\frac{\cos(\Delta x)-1}{\Delta x}+\cos(x)\frac{\sin(\Delta x)}{\Delta x}\Big)$

Now, by using the [limits sum & product rules](0. References.ipynb#Limits-rules)  we can see that this is the same as:

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \big(\sin(x)\frac{\cos(\Delta x)-1}{\Delta x}\big)+\lim\limits_{\Delta x \to 0} \big(\cos(x)\frac{\sin(\Delta x)}{\Delta x}\big)$

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \big(\sin(x)\big)\lim\limits_{\Delta x \to 0} \big(\frac{\cos(\Delta x)-1}{\Delta x}\big)+\lim\limits_{\Delta x \to 0} \big(\cos(x)\big)\lim\limits_{\Delta x \to 0} \big(\frac{\sin(\Delta x)}{\Delta x}\big)$

Now, consider the second limits in each of the product terms separately. For now, I'm just going to state their values...

$\lim\limits_{\Delta x \to 0}\frac{\cos(\Delta x)-1}{\Delta x}=0$

$\lim\limits_{\Delta x \to 0}\frac{\sin(\Delta x)}{\Delta x}=1$

You can read the geometrical proofs of these limits [here](2c. Some geometric proofs.ipynb), but maybe seeing it graphically might help if you're not convinced...

```octave
posx = [0.1:0.01:3];
negx = [-3:0.01:-0.1];

posy1 = (cos(posx)-1)./posx;
negy1 = (cos(negx)-1)./negx;

posy2 = sin(posx)./posx;
negy2 = sin(negx)./negx;

plot(posx,posy1,'LineWidth',3,'color','b', ...
     posx,posy2,'LineWidth',3,'color','r', ...
     negx,negy1,'LineWidth',3,'color','b', ...
     negx,negy2,'LineWidth',3,'color','r');

set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");

legend ({'(cos(x)-1)/x','sin(x)/x'});
legend boxoff;
```

So, using the limits, $\lim\limits_{\Delta x \to 0}\frac{\cos(\Delta x)-1}{\Delta x}=0$ and $\lim\limits_{\Delta x \to 0}\frac{\sin(\Delta x)}{\Delta x}=1$...

$\frac{d}{dx}\sin x = \lim\limits_{\Delta x \to 0} \big(\sin(x)\big)\lim\limits_{\Delta x \to 0} \big(\frac{\cos(\Delta x)-1}{\Delta x}\big)+\lim\limits_{\Delta x \to 0} \big(\cos(x)\big)\lim\limits_{\Delta x \to 0} \big(\frac{\sin(\Delta x)}{\Delta x}\big)$


$= \sin(x) \times 0+\cos(x) \times 1$

$= \cos(x)$

### cos

$\frac{d}{dx}\cos x = -\sin x$

**Proof:**

The proof is similar to the proof for $\frac{d}{dx}\sin x$

$\frac{d}{dx}\cos x = \lim\limits_{\Delta x \to 0} \frac{\cos(x+\Delta x)-\cos(x)}{\Delta x}$

Using the [angle sum & difference trigonometric identities](0. References.ipynb#Trigonometric-identities)...

$\cos(a+b)=\cos(a)\cos(b)-\sin(a)\sin(b)$

$\cos(x+\Delta x)=\cos(x)\cos(\Delta x)-\sin(x)\sin(\Delta x)$

Therefore:

$\frac{d}{dx}\cos x = \lim\limits_{\Delta x \to 0} \frac{\cos(x)\cos(\Delta x)-\sin(x)\sin(\Delta x)-\cos(x)}{\Delta x}$

Again, with some shuffling and factoring we can end up with the same fractions that we know go to 0 and 1 when we take the limits (see the $\sin$ proof above):

$\frac{d}{dx}\cos x = \lim\limits_{\Delta x \to 0} \Big(\frac{\cos(x)\cos(\Delta x)-\cos(x)}{\Delta x}-\frac{\sin(x)\sin(\Delta x)}{\Delta x}\Big)$

$\frac{d}{dx}\cos x = \lim\limits_{\Delta x \to 0} \Big(\cos(x)\frac{\cos(\Delta x)-1}{\Delta x}-\sin(x)\frac{\sin(\Delta x)}{\Delta x}\Big)$

So, using the limits, $\lim\limits_{\Delta x \to 0}\frac{\cos(\Delta x)-1}{\Delta x}=0$ and $\lim\limits_{\Delta x \to 0}\frac{\sin(\Delta x)}{\Delta x}=1$...

$\frac{d}{dx}\cos x = \lim\limits_{\Delta x \to 0} \Big(\cos(x)\frac{\cos(\Delta x)-1}{\Delta x}-\sin(x)\frac{\sin(\Delta x)}{\Delta x}\Big)$

$= \cos(x) \times 0 -\sin(x) \times 1$

$= -\sin(x)$

### tan

$\frac{d}{dx}\tan(x)=\sec^2 x$

**Proof:**

$\tan(x)=\frac{\sin(x)}{\cos(x)}$

...from the [basic trigonometric identities](0. References.ipynb#Trigonometric-identities). So we can use the quotient rule ([defined and proven later](#Quotient-rule)) to find the derivative...

$\frac{d}{dx}\Big(\frac{f(x)}{g(x)}\Big)=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x))^2}$

Therefore:

$\frac{d}{dx}\tan(x)=\frac{d}{dx}\Big(\frac{\sin(x)}{\cos(x)}\Big)=\frac{\cos(x)\cos(x)+\sin(x)\sin(x)}{\cos^2(x)}$

$=\frac{\cos^2(x)+\sin^2(x)}{\cos^2(x)}$

$\cos^2(x)+\sin^2(x)=1$ (from the [pythagorean trigonometric identities](0. References.ipynb#Trigonometric-identities)) so...

$\frac{d}{dx}\tan(x)=\frac{1}{\cos^2(x)}$

$=\Big(\frac{1}{\cos(x)}\Big)^2$

$\frac{1}{\cos(x)}=\sec(x)$ (from the [basic trigonometric identities](0. References.ipynb#Trigonometric-identities)) so...

$\frac{d}{dx}\tan(x)=\sec^2(x)$

(Note that the quotient rule can be used in the same way to find the derivative of the other trigonometric functions.)

## Product rule

$\frac{d}{dx}(f(x)g(x))=f'(x)g(x)+f(x)g'(x)$

**Proof:**

Now, this one gets a bit hairy so lets take it very slowly to make sure we're not doing anything illegal!

As with the other proofs, start with the limit of the difference quotient [definition of the derivative](1. Definition of Derivative.ipynb):

Let $h(x)=f(x)g(x)$, then...

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\frac{h(x+\Delta x)-h(x)}{\Delta x}$

$=\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x+\Delta x)-f(x)g(x)}{\Delta x}$

The trick is to think about what terms we'd expect to see for $f'(x)g(x)$ to appear:

$f'(x)g(x)=g(x)\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x}$

...using the limit of the difference quotient definition of the derivative again, this time for $f'(x)$.

$g(x)$ can be treated like a constant because it doesn't contain the limit variable, $\Delta x$. The [limit constant multiple rule](0. References.ipynb#Limits-rules) says that a constant times a limit of an expression is the same as the limit of the expression times the constant. That's a bit ambiguous to say in words without knowing where the brackets should be in the sentence. Here it is without the ambiguity...

$g(x)\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x}=\lim\limits_{\Delta x \to 0}\Big(g(x)\frac{f(x+\Delta x)-f(x)}{\Delta x}\Big)$

Now, expand out the product...

$f'(x)g(x)=\lim\limits_{\Delta x \to 0}\frac{g(x)(f(x+\Delta x)-f(x))}{\Delta x}$

$=\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x)-f(x)g(x)}{\Delta x}$

So, looking at the definition we've got so far...

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x+\Delta x)-f(x)g(x)}{\Delta x}$

...we can see we've got the $-f(x)g(x)$ term in the numerator already but not the $f(x+\Delta x)g(x)$ term. We could just add it and see what happens. (Of course, we have to subtract it as well so we're not changing the overall value of the expression!)

$f(x+\Delta x)g(x)-f(x+\Delta x)g(x)=0$ so adding it to the numerator doesn't change the value...

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x+\Delta x)-f(x)g(x)+f(x+\Delta x)g(x)-f(x+\Delta x)g(x)}{\Delta x}$

Rearrange the numerator a bit:

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x)-f(x)g(x)+f(x+\Delta x)g(x+\Delta x)-f(x+\Delta x)g(x)}{\Delta x}$

...and splitting it into the sum of two more managable fractions:

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\Big(\frac{f(x+\Delta x)g(x)-f(x)g(x)}{\Delta x}+\frac{f(x+\Delta x)g(x+\Delta x)-f(x+\Delta x)g(x)}{\Delta x}\Big)$

Then, the [limit sum rule](0. References.ipynb#Limits-rules) says the limit of a sum is the same as the sum of the limits...

$\frac{d}{dx}h(x)=\lim\limits_{\Delta x \to 0}\Big(\frac{f(x+\Delta x)g(x)-f(x)g(x)}{\Delta x}\Big)+\lim\limits_{\Delta x \to 0}\Big(\frac{f(x+\Delta x)g(x+\Delta x)-f(x+\Delta x)g(x)}{\Delta x}\Big)$

And the first limit is exactly what we're looking for. It's the same as what we got above for $f'(x)g(x)$.

$\frac{d}{dx}h(x)=f'(x)g(x)+\lim\limits_{\Delta x \to 0}\Big(\frac{f(x+\Delta x)g(x+\Delta x)-f(x+\Delta x)g(x)}{\Delta x}\Big)$

Now let's do some factoring of the second term:

$\frac{d}{dx}h(x)=f'(x)g(x)+\lim\limits_{\Delta x \to 0}\Big(f(x+\Delta x)\frac{(g(x+\Delta x)-g(x))}{\Delta x}\Big)$

And applying the [limit product rule](0. References.ipynb#Limits-rules) which says the limit of a product is the same as the product of the limits...

$\frac{d}{dx}h(x)=f'(x)g(x)+\lim\limits_{\Delta x \to 0}\big(f(x+\Delta x)\big)\lim\limits_{\Delta x \to 0}\Big(\frac{(g(x+\Delta x)-g(x))}{\Delta x}\Big)$

Oh, look! That last term is the definition of the derivative of $g$ with respect to $x$.

$\frac{d}{dx}h(x)=f'(x)g(x)+\lim\limits_{\Delta x \to 0}\big(f(x+\Delta x)\big)g'(x)$

Finally, if we take the limit as $\Delta x \to 0$ then $f(x+\Delta x)$ simply becomes $f(x)$...

$\frac{d}{dx}h(x)=f'(x)g(x)+f(x)g'(x)$

Which is what we set out to prove!

(**Note:** this assumes that both functions $f$ and $g$ are indeed differentiable, which also means they [must be continuous](1a. Differentiable functions must be continuous.ipynb). This is important to note because it gives us confidence that the way we handled the limits in the above proof is valid. In a continuous function any limit exists and is simply the value of the function at the point of the limit.)

## Quotient rule

$\frac{d}{dx}\Big(\frac{f(x)}{g(x)}\Big)=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x))^2}$

**Proof:**

$\frac{d}{dx}\Big(\frac{f(x)}{g(x)}\Big)=\lim\limits_{\Delta x \to 0}\frac{\frac{f(x+\Delta x)}{g(x+\Delta x)}-\frac{f(x)}{g(x)}}{\Delta x}$

Now that's a bit unwieldy, let's just concentrate on the numerator for a bit and see if we can simplify it:

$\frac{f(x+\Delta x)}{g(x+\Delta x)}-\frac{f(x)}{g(x)}$

Suppose we define a couple of new variables, $\Delta f$ and $\Delta g$, to represent the change in the respective functions as $x$ changes by $\Delta x$...

$\Delta f=f(x+\Delta x)-f(x)$

$\Delta g=g(x+\Delta x)-g(x)$

...which can be transposed to...

$f(x+\Delta x)=f(x)+\Delta f$

$g(x+\Delta x)=g(x)+\Delta g$

...and substituted into the numerator...

$\frac{f(x+\Delta x)}{g(x+\Delta x)}-\frac{f(x)}{g(x)}$

$=\frac{f(x)+\Delta f}{g(x)+\Delta g}-\frac{f(x)}{g(x)}$

$=\frac{(f(x)+\Delta f)g(x)-f(x)(g(x)+\Delta g)}{(g(x)+\Delta g)g(x)}$

$=\frac{f(x)g(x)+(\Delta f)g(x)-f(x)g(x)-f(x)(\Delta g))}{(g(x)+\Delta g)g(x)}$

$=\frac{(\Delta f)g(x)-f(x)(\Delta g))}{(g(x)+\Delta g)g(x)}$

Now put that back into the difference quotient:

$\frac{\frac{f(x+\Delta x)}{g(x+\Delta x)}-\frac{f(x)}{g(x)}}{\Delta x}=\frac{\frac{(\Delta f)g(x)-f(x)(\Delta g))}{(g(x)+\Delta g)g(x)}}{\Delta x}=\frac{1}{\Delta x}\frac{(\Delta f)g(x)-f(x)(\Delta g))}{(g(x)+\Delta g)g(x)}$

...which can be rearranged as...

$\frac{1}{\Delta x}\frac{(\Delta f)g(x)-f(x)(\Delta g))}{(g(x)+\Delta g)g(x)}=\frac{\frac{\Delta f}{\Delta x}g(x)-f(x)\frac{\Delta g}{\Delta x}}{(g(x)+\Delta g)g(x)}$

Notice that if we take $\frac{\Delta f}{\Delta x}$ and substitute $\Delta f$ with its original definition...

$\frac{\Delta f}{\Delta x}=\frac{f(x+\Delta x)-f(x)}{\Delta x}$

...then we get the definition of the derivative of $f$ as $\Delta x \to 0$. (And similarly, $\frac{\Delta g}{\Delta x}$ gives us the derivative of $g$ as $\Delta x \to 0$.) Therefore,

$\frac{\frac{\Delta f}{\Delta x}g(x)-f(x)\frac{\Delta g}{\Delta x}}{(g(x)+\Delta g)g(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x)+\Delta g)g(x)}$

Now substitute for $\Delta g$ in the denominator:

$\frac{\frac{\Delta f}{\Delta x}g(x)-f(x)\frac{\Delta g}{\Delta x}}{(g(x)+\Delta g)g(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x)+g(x+\Delta x)-g(x))g(x)}$

You can see that when you take the limit $\Delta x \to 0$ then $g(x+\Delta x)=g(x)$. (We know that $g$ must be differentiable and therefore continuous so this is valid.)

$\lim\limits_{\Delta x \to 0}\frac{f'(x)g(x)-f(x)g'(x)}{(g(x)+g(x+\Delta x)-g(x))g(x)}=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x)+g(x)-g(x))g(x)}$

$=\frac{f'(x)g(x)-f(x)g'(x)}{(g(x))^2}$

## Chain rule

$\frac{d}{dx}f(g(x))=f'(g(x))g'(x)$

**Proof:**

Introducing a couple of variables will make things much more managable:

$y=f(g(x))$

$z=g(x)$

Therefore,

$y=f(z)$

...and the derivative we're trying to find is $\frac{dy}{dx}$.

We know from the [definition of a derivative](1. Definition of Derivative.ipynb) that:

$\frac{dy}{dx}=\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta x}$

We can multiply by $\Delta z$ and divide by $\Delta z$ without changing anything, so we can rewrite this as...

$\frac{dy}{dx}=\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta x}=\lim\limits_{\Delta x \to 0}\bigg(\frac{\Delta y}{\Delta z}\frac{\Delta z}{\Delta x}\bigg)$

Now, we're assuming that each function, $f$ and $g$, are differentiable at $(x)$, which means they must also be continuous at $x$. This makes it easier to deal with the limit. We can say the the limit of a product is equal to the product of two limits...

$\lim\limits_{\Delta x \to 0}\bigg(\frac{\Delta y}{\Delta z}\frac{\Delta z}{\Delta x}\bigg)=\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta z}\lim\limits_{\Delta x \to 0}\frac{\Delta z}{\Delta x}$

It's easy to see from the [definition of a derivative](1. Definition of Derivative.ipynb) that the second limit is the derivative of $g(x)$:

$\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta z}\lim\limits_{\Delta x \to 0}\frac{\Delta z}{\Delta x}=\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta z}g'(x)$

Now remember that we're assuming that $z=g(x)$ is differentiable and continuous at $x$. This means that as $\Delta x \to 0$ then $\Delta z \to 0$ also. So...

$\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta z}=\lim\limits_{\Delta z \to 0}\frac{\Delta y}{\Delta z}$

...which gives us the definition of the derivative of $y$ with respect to $z$, and (because $z=g(x)$ and $y=f(g(x))$), the derivative of $y$ with respect to $z$ is the same as $f'(g(x))$.

$\lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta z}g'(x)=f'(g(x))g'(x)$
