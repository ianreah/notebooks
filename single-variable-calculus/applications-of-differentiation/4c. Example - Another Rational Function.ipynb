*Is there anywhere the function is undefined? If it's a log or square-root function then the input must be positive. If it's a rational function then the denominator can't be 0 (this creates vertical asymptotes).*

A rational function is undefined when the denominator is 0 but our denominator can never be 0. The square of a real number is always positive so $1+ x^2 \gt 1$.

---

*Consider the "end behaviour" (as $x \to \infty$ and $x \to -\infty$). This creates horizontal asymptotes.*

As $x \to \infty$ the higher power in the denominator wins out, i.e., a large value on the numerator but a much larger number on the denominator means the overall value is getting smaller and tending towards $0$.

Similarly, as $x \to -\infty$ we have a large negative value on the numerator with a much larger positive calue on the denominator meaning the overall value is negative but again tending towards $0$.

So we have a horizontal asymptote at $0$.

---

*Look at the sign of the function and note where its sign changes.*

The sign of a rational function is postive when the numerator and denominator are the same sign and negative when they're different. We've already determined that the denominator is always positive so, in this case, the sign of the function is governed by the sign of the numerator.

The function is positive when $x \gt 0$ and negative when $x \lt 0$.

![](img\sketching-example-3b.jpg)

---

*Using information from the first derivative determine where the function is increasing/decreasing. Places where the first derivative is 0 give maximum or minimum values of the function.*

$f(x)=\frac{x}{1+x^2}$

$f'(x)=\frac{\frac{d}{dx}(x)(1+x^2)-x\frac{d}{dx}(1+x^2)}{(1+x^2)^2}$

$=\frac{1+x^2-2x^2}{(1+x^2)^2}$

$=\frac{1-x^2}{(1+x^2)^2}$

Just like the original function we've got a rational with a denominator that's always positive, so the sign of the derivative is determined by the sign of the numerator, $1-x^2$.

This is positive when $1 \gt x^2$ (i.e., $-1 \lt x \lt 1$), and negative everywhere else. This tells us that the function is increasing when $-1 \lt x \lt 1$ and decreasing everywhere else. The minimum of the function must be at $x=-1$ and the maximum must be when $x=1$

Minimum: $f(-1)=\frac{-1}{1+(-1)^2}=-\frac{1}{2}$

Maximum: $f(1)=\frac{1}{1+1^2}=\frac{1}{2}$

(Notice I've changed the scale of the y-axis now that we know more about the range of the function.)

---

*Use information from the second derivative to determine the shape of the curve between the critical points. Also, consider where the second derivative is 0 to determine where the shape changes.*

$f'(x)=\frac{1-x^2}{(1+x^2)^2}$

$=(1-x^2)(1+x^2)^{-2}$

$g(x)=1-x^2$

$h(x)=(1+x^2)^{-2}$

$f'(x)=g(x)h(x)$

$f''(x)=g'(x)h(x)+g(x)h'(x)$

$g'(x)=-2x$

$h'(x)=-2(1+x^2)^{-3}.2x$

$=-4x(1+x^2)^{-3}$

$f''(x)=-2x(1+x^2)^{-2}+(1-x^2)(-4x)(1+x^2)^{-3}$

$=\frac{-2x}{(1+x^2)^2}+\frac{-4x(1-x^2)}{(1+x^2)^3}$

$=\frac{-2x(1+x^2)-4x(1-x^2)}{(1+x^2)^3}$

$=\frac{-2x-2x^3-4x+4x^3}{(1+x^2)^3}$

$=\frac{-6x+2x^3}{(1+x^2)^3}$

$=\frac{2x(x^2-3)}{(1+x^2)^3}$

Again, the denominator is always positive so the sign of the second derivative is determined by the sign of the numerator. In the numerator we have a product of two terms ($2x$ and $x^2-3$) so the result is positive when the sign of the two terms is the same and negative when they're different.

$2x$ is positive when $x \gt 0$.

$x^2-3$ is positive when $x^2 \gt 3$, or $x \gt \pm\sqrt{3}$, or $-\sqrt{3} \lt x \lt \sqrt{3}$

*Consider any other "important" points and plot them (e.g., the behaviour around any asymptotes)*

Well we've got that horizontal asymptote at $0$. We also know that the function is positive when $x \gt 0$ and that the function is concave down when $x \gt \sqrt{3}$. So we know that as $x \to \infty$ the function approaches $0$ from the positive side in a concave down curve. Similarly, when $x \to -\infty$ it'll approach $0$ from the negative side in a concave down curve.

...so the function is odd. So we need to make sure everything we've done so far and what we end up with has rotational symmetry with respect to the origin.

*Where (if anywhere) does the curve cross the x-axis or y-axis?*

$\frac{x}{1+x^2}=0$ only when $x=0$. So the function passes through the origin and then never touches either axis again.
