Related Rates problems are problems where multiple quantities are changing but there is some relationship between the different rates of change. Typically the rate of change of one quantity will be known and the problem involves calculating the rate of change of another quantity.

**Example:** Police are 30 feet from the side of the road. Their radar sees your car approaching at 80 feet per second when your car is 50 feet away from the radar gun. The speed limit is 65 miles per hour (which translates to 95 feet per second). Are you speeding?

![](img\related-rates-example-1.png)

The first step is to make sure we understand clearly what quantities are changing and what are constant. The police car isn't moving so its perpendicular distance from the road is constant, $30$. Our car is moving along the road but notice how this gives us two changing distance...

1. The distance along the road. Let's call this $x$ (where $x=0$ when it is in line with police car).
1. The distance that the radar measures. Let's call that $D$. 

Next, let's consider what we know about the rates of change of these.

1. The rate of change of $x$ is basically the speed of the car. We don't know this; that's what we need to find out.
1. The rate of change of $D$ is given. It's the speed that the radar measures: 80 feet per second. (Or, strictly speaking the rate of change is $-80$ feet per second as the distance that the radar sees is getting smaller.)

The final piece of the puzzle is to work out what relationships we can derive about the problem. This example is a pretty much a right-angled triangle so we can use pythagoras to define a relationship between the two changing distances...

$D^2=x^2+30^2$

To solve the problem we need to differentiate to find the appropriate rate of change. But before we do that it's important to be clear about what we're differentiating with respect to. Well, the distances are changing with respect to time so we should think of $x$ and $D$ as functions of time, $t$. This means we should use implicit differentiation to differentiate with respect to $t$...

$D^2=x^2+30^2$

$2D\frac{dD}{dt}=2x\frac{dx}{dt}$

Rearrange to solve for the rate of change we need to find, $\frac{dx}{dt}$...

$\frac{2D\frac{dD}{dt}}{2x}=\frac{dx}{dt}$

$\frac{dx}{dt}=\frac{2D}{2x}\frac{dD}{dt}$

$\frac{dx}{dt}=\frac{D}{x}\frac{dD}{dt}$

We're given $\frac{dD}{dt}$ (-80 feet per second). We're given a value for $D$ at a particular instant in time (50 feet), and we can calculate $x$ at the same instant in time. (It's a right-angled triangle and conveniently it makes a 3-4-5 triangle at that instant.)

$\frac{dx}{dt}=\frac{50}{40}(-80)=100$ feet per second

...oh dear! We're speeding.

## More changing variables

The above example contained two rates of change. One was given and we used the relationship between the varying quantities to calculate the other. What if there is more than two variables changing? Well, just like with [this optimisation problem example](5a.%20Optimisation%20problems%20and%20implicit%20differentiation.ipynb), you can either reduce the number of variables first by re-arranging & substitution or you can go straight to implicit differentation.

**Example:** Consider a conical tank whose radius at the top is 4 feet and whose depth is 10 feet. Itâ€™s being filled with water at the rate of 2 cubic feet per minute. How fast is the water level rising when it is at depth 5 feet?

![](img\related-rates-example-2.png)

The first step is to make sure we understand clearly what quantities are changing and what are constant. There are **3** changing quantities...

1. The volume of water in the cone. We'll call this $V$.
1. The height of the water level. This is $h$ in the above diagram.
1. The radius of the cone at the water level height. This is $r$ in the above diagram.

Next, let's consider what we know about the rates of change of these. All 3 are changing *with respect to time*.

1. We're given the rate of change of $V$. $\frac{dV}{dt}=2$ feet per minute.
1. The rate of change of $h$, (i.e., $\frac{dh}{dt}$), is what we're asked to find out (when the water is at depth 5 feet).
1. The rate of change of $r$, (i.e., $\frac{dr}{dt}$), is also unknown - potentially making this problem a bit harder than the previous example. So we need a strategy for dealing with this (either remove it via rearranging and substitution or go straight to implicit differentiation and deal with it later.) The former is probably a bit simpler in this case...

$\frac{r}{h}=\frac{4}{10}$ (similar triangles)

$r=\frac{2}{5}h$

The volume of a cone is:

$V=\frac{1}{3}\pi r^2h$

$V=\frac{1}{3}\pi ({\frac{2}{5}h})^2h$ (subsitution to get rid of $r$)

$V=\frac{1}{3}\pi ({\frac{2}{5}h})({\frac{2}{5}h})h$

$V=\frac{4}{75}\pi h^3$

Differentiate with respect to $t$:

$\frac{dV}{dt}=\frac{4}{75}\pi 3h^2\frac{dh}{dt}$

$\frac{dV}{dt}=\frac{12}{75}\pi h^2\frac{dh}{dt}$

$\frac{dV}{dt}=\frac{4}{25}\pi h^2\frac{dh}{dt}$

Plug in the values we know (at a particular point in time):

$2=\frac{4}{25}\pi 5^2\frac{dh}{dt}$

$2=\frac{100}{25}\pi\frac{dh}{dt}$

$2=4\pi\frac{dh}{dt}$

$\frac{2}{4\pi}=\frac{dh}{dt}$

$\frac{dh}{dt}=\frac{1}{2\pi}$
