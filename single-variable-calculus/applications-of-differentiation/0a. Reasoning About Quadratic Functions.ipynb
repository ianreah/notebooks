A linear function has a constant slope, or, a constant rate of change. It's the $m$ in $y=mx+c$, and the result of taking the first derivative.

$y=mx+c$

$y'=m$

It's easy to reason about the effects of the coefficients on the line. A higher $m$ gives the line a steeper slope; a negative $m$ gives the line a downward slope (as $x$ increases); the constant $c$ simply moves the whole line up or down; vertically.

Can we simlarly describe the behaviour of a quadratic function in terms of its coefficients?

$y=ax^2+bx+c$

First, familiarise yourself with what the function looks like in its simplest form, $y=x^2$, i.e., $a=1$, $b=0$, $c=0$.

```octave
x=[-5:0.01:5];
y=x.^2;

plot(x,y,'LineWidth', 2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

Clearly a quadratic function doesn't have a constant slope like the linear function, but the rate of change of the slope is constant. (The rate of change of the rate of change is constant! :mindblown:)

We know it's a constant rate of change from the derivative. If you take the derivative of a quadratic function you get a linear function. This describes the slope (rate of change) of the quadratic function at each point. And, as we've already said, the linear function has a constant rate of change.

Again, the rate of change of the quadratic function isn't constant - but it changes at a constant rate defined by the linear function you get from its derivative. If you take the second derivative you get the slope of the rate of change line!

$y=ax^2+bx+c$

$y'=2ax+b$

$y''=2a$

Or, as a shortcut, you can say the slope of the rate of change is twice the coefficient of the $x^2$ term, i.e., $2a$.

The coefficient, $a$ of the $x^2$ term therefore effects the rate of change of the slope of the curve. A higher $a$ gives a narrower, steeper curve; a lower $a$ gives a fatter, shallower curve; a negative $a$ inverts the curve.

```octave
x=[-5:0.01:5];

# a=1
y1=x.^2;

# a=2
y2=2.*x.^2;

# a=0.5;
y3=0.5.*x.^2;

plot(x,y1,'LineWidth', 2, x,y2, 'LineWidth', 2, x,y3, 'LineWidth', 2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");

legend ({'y=x^2','y=2x^2','y=0.5x^2'});
legend boxoff;
```

```octave
x=[-5:0.01:5];
y=-2.*x.^2;

plot(x,y,'LineWidth', 2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");

legend ({'y=-2x^2'});
legend boxoff;
```

The effect of the $b$ coefficient is a bit trickier to reason about. First, let's go back to this idea that the derivative of the quadratic function gives a linear function that defines the rate of change (or slope) of the original function at each point.

$y=ax^2+bx+c$

$y'=2ax+b$

So, if $2a$ is the slope of the linear function ($m$ in our original description of the straight line), then $b$ is the constant offset. In other words the value that makes the straight line move up or down vertically.

Now, if this straight line defines the rate of change of the quadratic function at each point, then the point where the line crosses the x-axis (i.e., where its value is $0$) must be the x-value of the "apex" of the curve (i.e., the point where gradient of the curve is $0$).

A higher $b$ moves the line up, which means it crosses the x-axis further to the left, so the apex of the quadratic curve will be further to the left, and vice versa.

```octave
x=[-5:0.01:5];

# b=0
y1=x.^2;

# b=3
y2=x.^2+3.*x;

# b=-3;
y3=x.^2-3.*x;

y1_prime=2.*x;
y2_prime=2.*x+3;
y3_prime=2.*x-3;

plot(x,y1, 'LineWidth', 2, 'color', 'blue', ...
     x,y2, 'LineWidth', 2, 'color', 'red', ...
     x,y3, 'LineWidth', 2, 'color', 'green', ...
     x,y1_prime, 'color', 'blue', ...
     x,y2_prime, 'color', 'red', ...
     x,y3_prime, 'color', 'green');
     
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");

legend ({'y=x^2','y=x^2+3x','y=x^2-3x'});
legend boxoff;
```

The graph above is a bit busy, but hopefully you can see how the $b$ coefficient results in the gradient (the linear function from the derivate of the quadratic function) moving up/down as you'd expect from the constant coefficient. And that the apex of the quadratic function lines up with where the relevant gradient line crosses the x-axis (i.e., where the gradient of the quadratic curve is 0).

But, notice that the quadratic function doesn't only move left or right as we change $b$, but also up/down as well. Why is that?

To understand this, let's think about the value of the function at that apex point.

A higher $b$ value causes the apex point to move to the left as shown above, i.e., the apex is at a negative $x$ value. For the above example with $b=3$, the $x$ value where the gradient is $0$ is:

$2x+3=0$

$2x=-3$

$x=-\frac{3}{2}$

The value of the quadratic function when $x=-\frac{3}{2}$ is:

$y=x^2+bx$

$=(-\frac{3}{2})^2+b(-\frac{3}{2})$ ...and $b=3$ so...

$=(-\frac{3}{2})^2+3(-\frac{3}{2})$

$=\frac{9}{4}-\frac{9}{2}=-\frac{9}{4}$

Similarly, for lower $b$ values, the apex point moves to the right but also down. For the above example with $b=-3$, the value where the gradient is $0$ becomes:

$2x-3=0$

$2x=3$

$x=\frac{3}{2}$

And, the value of the quadratic function when $x=\frac{3}{2}$ is:

$y=x^2+bx$

$=(\frac{3}{2})^2+b(\frac{3}{2})$ ...and $b=-3$ so...

$=(\frac{3}{2})^2-3(\frac{3}{2})$

$=\frac{9}{4}-\frac{9}{2}=-\frac{9}{4}$

In fact, the apex actually follows an inverted curve...

```octave
x=[-5:0.01:5];
y=x.^2;

b=-2.*x;
apex=x.^2+b.*x;

example=x.^2+5.*x;

plot(x,y, 'color', 'blue', 'LineWidth', 2, ...
     x,apex, 'color', 'blue', ...
     x,example, 'color', 'red');
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");

legend ({'original quadratic function','apex path as b changes','example when b=5'});
legend boxoff;
```

As you'd expect, the third coefficient $c$ simply adds a constant to the values so moves the curve up/down.
