*Is there anywhere the function is undefined? If it's a log or square-root function then the input must be positive. If it's a rational function then the denominator can't be 0 (this creates vertical asymptotes).*

$\ln$ is only defined for positive values. Because it's a rational function it's also undefined when the denominator is $0$, i.e., $\ln x=0$. This happens when $x=1$

So $x>0$ and we have a vertical asymptote at $x=1$.

![](img\sketching-example-4b.jpg)

---

*Consider the "end behaviour" (as $x \to \infty$ and $x \to -\infty$). This creates horizontal asymptotes.*

In this case the end behaviour is as $x \to \infty$ and $x \to 0$ (from the right).

As $x \to 0$ from the right then we get...

$f(0^+)=\frac{0^+}{\ln 0^+}$

The denominator is going to $-\infty$ but seeing as the numerator is going to $0$ then the result of the function also goes to $0$.

It's a bit trickier to get your head round what happens as $x \to \infty$. Start by plugging a "big" number and see how that behaves, say $x=10^{10}$...

$f(x)=\frac{x}{\ln x}$

$f(10^{10})=\frac{10^{10}}{\ln (10^{10})}$

$=\frac{10^{10}}{10 \ln 10}$

$=\frac{10^{9}}{\ln 10}$

The numerator is much larger than the denominator and as $x$ increases the numerator gets larger faster (imagine plugging in $x=10^{20}$ and consider how much the numerator has increased compared to the denominator!), so the function goes to $\infty$ as $x \to \infty$.

---

*Look at the sign of the function and note where its sign changes.*

The numerator is always positive so the sign of the function is determined by the sign of the denominator, $\ln x$. $\ln x$ is negative when $0 \lt x \lt 1$, and positive when $x > 1$.

![](img\sketching-example-4c.jpg)

---

*Using information from the first derivative determine where the function is increasing/decreasing. Places where the first derivative is 0 give maximum or minimum values of the function.*

$f(x)=\frac{x}{\ln x}$

$f'(x)=\frac{\ln x-x.(\frac{1}{x})}{(\ln x)^2}$

$=\frac{\ln x-1}{(\ln x)^2}$

Clearly the denominator is always positive. The numerator will be negative when $\ln x \lt 1$ and positive when $\ln x \gt 1$.

$\ln x=1$ when $x=e$ ($\ln$ is the inverse function of $x=e^y$, i.e., $\ln x=y$. If $\ln x=1$ then $x=e^1=e$)

$\ln x \lt 1$ when $x \lt e$

$\ln x \gt 1$ when $x \gt e$

The derivative is positive so the function will be increasing when $x \gt e$.
It's negative so the function will be decreasing when $x \gt e$.
Like the original function though, the second derivative is undefined when the denominator is $0$ so we still have that asymptote around $x=1$ (when $(\ln x)^2=0$) where strictly speaking we can't say the function is increasing or decreasing!

So, the gradient is zero when $x=e$. What's the value at this point?

$f(e)=\frac{e}{\ln e}$

$f(e)=e$ ($\ln e=1$)

The point $(e, e)$ is a critical point where the graph changes from decreasing to increasing.

![](img\sketching-example-4d.jpg)

---

*Use information from the second derivative to determine the shape of the curve between the critical points. Also, consider where the second derivative is 0 to determine where the shape changes.*

$f'(x)=\frac{\ln x-1}{(\ln x)^2}$

Let's treat the numerator & denominator separately first to help differentiate.

$n(x)=\ln x-1$

$n'(x)=\frac{1}{x}$

$d(x)=(\ln x)^2$

$d'(x)=2(\ln x)\frac{1}{x}$

$=\frac{2(\ln x)}{x}$

$f'(x)=\frac{n(x)}{d(x)}$

$f''(x)=\frac{n'(x)d(x)-n(x)d'(x)}{(d(x))^2}$

$=\frac{(\frac{1}{x})(\ln x)^2-(\ln x-1)(\frac{2(\ln x)}{x})}{((\ln x)^2)^2}$

$=\frac{(\frac{(\ln x)^2}{x})-(\frac{2(\ln x)(\ln x-1)}{x})}{((\ln x)^2)^2}$

$=\frac{(\ln x)^2-2(\ln x)(\ln x-1)}{x(\ln x)^4}$

$=\frac{(\ln x)((\ln x)-2(\ln x-1))}{x(\ln x)^4}$

$=\frac{(\ln x)-2(\ln x-1)}{x(\ln x)^3}$

$=\frac{\ln x-2\ln x+2}{x(\ln x)^3}$

$=\frac{2-\ln x}{x(\ln x)^3}$

* The second derivative is positive when the sign of the numerator & denominator are the same. It's negative when the signs are different.
  * **Numerator:** positive when $2 \gt \ln x$ (i.e., when $x \lt e^2$)
  * **Denominator:** $x$ is always positive so we only need to worry about the sign of $\ln x$. $\ln x$ is negative when $0 \lt x \lt 1$, and positive when $x > 1$.
* It's undefined when the denominator is $0$
* It's $0$ when the numerator is $0$

*Consider any other "important" points and plot them (e.g., the behaviour around any asymptotes)*

I think we've pretty much got it covered now, but just as a check let's consider the behaviour around the vertical asymptote at $x=1$.

When $x \to 1$ from the right (i.e., $x \gt 1$) then $\ln x$ is a very small positive number.

$\lim\limits_{x \to 1^+}\frac{x}{\ln x}=\frac{1^+}{0+}=\infty$

When $x \to 1$ from the right (i.e., $x \lt 1$) then $\ln x$ is a very small negative number.

$\lim\limits_{x \to 1^-}\frac{x}{\ln x}=\frac{1^-}{0-}=-\infty$
