Suppose we had quite a complicated function...

$f(x)=\frac{e^{-3x}}{\sqrt{1+x}}$

...and instead of having to calculate some exact values we wanted to find the linear approximation when $x\approx 0$ (assuming all the values we want to find are close enough to $0$)...

Well, we could differentiate to find $f'$ and calculate $f(0)$ and $f'(0)$, but the whole purpose of using linear approximation is that it's much easier and quicker to get to the calculations. So, instead we can find the linear approximation by algebraically combining the linear approximations we already have.

Let's remind ourselves of what they are:

* $\sin(x)\approx x$
* $\cos(x)\approx 1$
* $e^x\approx x+1$
* $\ln(1+x)\approx x$
* $(1+x)^r\approx rx+1$

...**Note:** these are all approximations when $x\approx 0$.

If we re-arrange our initial function it should be clearer which aproximations we can use...

$f(x)=\frac{e^{-3x}}{\sqrt{1+x}}$

$f(x)=e^{-3x}(1+x)^{-\frac{1}{2}}$

We can use $e^x\approx x+1$ substituting $x$ for $-3x$ and $(1+x)^r\approx rx+1$ 
with $r=-\frac{1}{2}$...

$f(x)\approx (1-3x)(1-\frac{1}{2}x)$

$\approx 1-3x-\frac{1}{2}x+\frac{3}{2}x^2$

Now before we go any further let's remind ourselved why we're doing this: it's an *approximation* to make it *easier and quicker* to get to the result. Well, there's two important points to remember that will make this easier and quicker...

* $x$ is close to $0$, therefore $x^2$ (and indeed any other higher degree terms if there were any) must be very close to $0$
* Also, we're dealing with a **linear** approximation - we've already ignored a bunch of higher degree terms in the initial approximations

...so it's ok to simply ignore any higher degree terms.

$f(x)\approx 1-3x-\frac{1}{2}x$

$\approx 1-\frac{7}{2}x$


```octave
x=[-0.2:0.01:0.2];
y=exp(-3.*x)./sqrt(1+x);
y_approx=1-7/2*x;

plot(x,y,'LineWidth',2,x,y_approx);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```
