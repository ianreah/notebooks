Did that re-arrangement of the mean value theorem on the last page remind you of anything?

$f(b)=f(a)+f'(c)(b-a)$

Maybe the formula we used for linear approximation?

$f(x)\approx f(x_0)+f'(x_0)(x-x_0)$

So, if we wanted to estimate $f(b)$ using the linear approximation at point $a$ we could write it as...

$f(b)\approx f(a)+f'(a)(b-a)$

## What's the difference between the mean value theorem and linear approximation?

Let's look closely at the differences between this and the re-arrangement of the mean value theorem. There are two differences:

1. The second term in the mean value theorem uses the derivative at some point $c$ rather than at the point $a$. Now, we don't know where point $c$ is, other than it exists somewhere between $b$ and $a$, but...
2. ...with the above change the approximation becomes an equality!

The linear approximation formula is based on the assumption that the slope between $a$ and $b$ is *approximately* equal to the slope of the function at $a$, i.e., $f'(a)$...

$f(b)\approx f(a)+f'(a)(b-a)$

$f(b)-f(a) \approx f'(a)(b-a)$

$\frac{f(b)-f(a)}{b-a} \approx f'(a)$

...whereas, the mean value theorem tells us that the slope between $a$ and $b$ is *exactly* equal to the slope of the function at some (unknown) point $c$, i.e., $f'(c)$ where $a \lt c \lt b$...

$f(b)=f(a)+f'(c)(b-a)$

$f(b)-f(a)=f'(c)(b-a)$

$\frac{f(b)-f(a)}{b-a}=f'(c)$

And, if the slope between $a$ and $b$ is a value of the derivative of the function between $a$ and $b$ then it must be between the maximum and minimum values of the derivative of the function between $a$ and $b$...

$\min\limits_{a \le x \le b} f'(x) \le f'(c) \le \max\limits_{a \le x \le b} f'(x)$

This assures us that the linear approximation gives a reasonable approximation when $a$ and $b$ are close.


## The mean value theorem and approximation error

Another way to look at this is to say that the difference between $f(a)$ and $f(b)$ is given by the last term in the linear approximation but using the derivative at some point $c$ rather than $a$...

$f(b)-f(a)=f'(c)(b-a)$

Think of $f(a)$ as a $0^{th}$ degree approximation of $f(b)$. So we're actually saying that the difference between the $0^{th}$ degree approximation and the actual value of $f(b)$ (i.e., the error of the approximation) is given by the last term in the $1^{st}$ degree approximation but using the derivative at some point $c$ instead.

Now this can be generalised to *higher degree* approximations. I don't know how to prove it just yet, but extending this to linear approximation, it's error would be given by the term we'd add for a *quadratic approximation* but using the second deriviative at some (unknown) point $c$...

$\frac{f''(c)}{2}(b-a)^2$

![](img\linear-approximation-error.png)

We didn't actually use the name previously, but the general formula we derived for an $n^{th}$ degree approximation is called the [Taylor series](https://en.wikipedia.org/wiki/Taylor_series):

$f(x)\approx f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2}(x-x_0)^2+...+\frac{f^{(n-1)}(x_0)}{(n-1)!}(x-x_0)^{n-1}+\frac{f^{(n)}(x_0)}{n!}(x-x_0)^n$

So, in general, the error of an $n^{th}$ degree approximation is based on the $(n+1)^{th}$ term in the Taylor series...

$\frac{f^{(n+1)}(c)}{(n+1)!}(x-x_0)^{n+1}$

We might not know how to prove that but we can test it out with a simple example. Consider a cubic polynomial...

$f(x)=x^3+2x+1$

...and a quadratic approximation of $f(3)$ at $x_0=1$:

$f(3)\approx f(1)+f'(1)(3-1)+\frac{f''(1)}{2}(3-1)^2$

$f'(x)=3x^2+2$

$f''(x)=6x$

So...

$f(3)\approx 4+5(3-1)+\frac{6}{2}(3-1)^2\approx 4+10+12\approx 26$

And the error of this approximation will be based on the next term in the Taylor series...

$\frac{f^{(3)}(c)}{3!}(3-1)^3$

Now, it's true that we don't know what $c$ is but when we differentiate again to get $f^{(3)}$ we realise it's constant so it doesn't matter that we don't know $c$ (in this example)...

$f^{(3)}(x)=6$

$\frac{f^{(3)}(c)}{3!}(3-1)^3=\frac{6}{6}(3-1)^3=8$

The *actual* value of $f(3)$ is...

$f(3)=3^3+2(3)+1=34$

...so the difference between our approximation and the actual value is $8$ which matches the error calculation from the Taylor series!
