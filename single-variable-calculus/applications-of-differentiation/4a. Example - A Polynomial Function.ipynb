*Is there anywhere the function is undefined? If it's a log or square-root function then the input must be positive. If it's a rational function then the denominator can't be 0 (this creates vertical asymptotes).*

This function is defined for all values of $x$.

---

*Consider the "end behaviour" (as $x \to \infty$ and $x \to -\infty$). This creates horizontal asymptotes.*

* As $x \to \infty$ then $x^3$ grows much more rapidly than $3x$, so $f(x) \approx -x^3 \to -\infty$.
* Simlarly, as $x \to -\infty$ then $f(x) \approx x^3 \to \infty$.

![](img\sketch-end-behaviour.png)

---

*Look at the sign of the function and note where its sign changes.*

Let's look at the points where the function is zero and consider the sign either side of these points...

$3x-x^3=0$

$x(3-x^2)=0$

$x=0$ or $3-x^2=0$

$x=0$ or $x^2=3$

$x=-\sqrt{3}$ or $x=0$ or $x=\sqrt{3}$

* For $x \lt -\sqrt{3}$ the $x^3$ term gets smaller faster than the $3x$ term so $3x \gt x^3$ resulting in a positive value
* For $-\sqrt{3} \lt x \lt 0$, $x^3 \gt 3x$ resulting in a negative value
* For $0 \lt x \lt \sqrt{3}$, $x^3 \lt 3x$ resulting in a positive value
* For $x \gt \sqrt{3}$ the $x^3$ term gets larger faster than the $3x$ term so $x^3 \gt 3x$ resulting in a negative value

![](img\sketch-sign-changes.png)

---

*Using information from the first derivative determine where the function is increasing/decreasing. Places where the first derivative is 0 give maximum or minimum values of the function.*

$f'(x)=3-3x^2$

The first derivative is positive when $3x^2 \lt 3$ (i.e., when $x^2 \lt 1$, so $-1 \lt x \lt +1$) and negative when $3x^2 \gt 3$ (i.e., everywhere else). This tells us that $f$ is increasing between $-1$ and $1$ and decreasing outside this range.

The values when $x=-1$ and $x=1$ gives us maximum and minimum values of the function.

$f(1)=3(1)-1^3=2$

$f(-1)=3(-1)-(-1)^3=-2$

![](img\sketch-max-min.png)

---

*Use information from the second derivative to determine the shape of the curve between the critical points. Also, consider where the second derivative is 0 to determine where the shape changes.*

$f''(x)=-6x$

The second derivative is positive when $x \lt 0$ and negative when $x \gt 0$. This tells us that $f$ is *concave up* when $x \lt 0$ and *concave down* when $x \gt 0$. It changes from concave up to down when $x=0$ (think of this as the *critical point* in the derivative). Also, the value of the function at this point is $f(0)=3(0)-0^3=0$

![](img\sketch-second-derivative.png)

*Is the function [even/odd](0.%20References.ipynb#Determine-whether-a-function-is-odd-or-even-%28or-neither%29)? Make sure what you've drawn so far fits with this.*

$f(x)=3x-x^3$

$-f(x)=-(3x-x^3)$

$=-3x+x^3$

$f(-x)=3(-x)-(-x)^3$

$=-3x+x^3$

$f(-x)=-f(x)$ ...therefore, this function is ODD. It has rotational symmetry about the origin. Luckily, everything we've drawn so far does also!
