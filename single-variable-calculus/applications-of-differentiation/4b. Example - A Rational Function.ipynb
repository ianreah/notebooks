*Is there anywhere the function is undefined? If it's a log or square-root function then the input must be positive. If it's a rational function then the denominator can't be 0 (this creates vertical asymptotes).*

The function is undefined when the denominator is 0, i.e., when $x+2=0$ or $x=-2$. This creates a vertical asymptote at $x=-2$. Let's consider the left and right limits as $x \to -2$.

**Right limit:**

$\lim\limits_{x \to -2^+}f(x)=\frac{(-2)^++1}{(-2)^++2}=\frac{(-1)^+}{0^+}=-\infty$ (Remember: $\frac{-}{+}=-$)

**Left limit:**

$\lim\limits_{x \to -2^-}f(x)=\frac{(-2)^-+1}{(-2)^-+2}=\frac{(-1)^-}{0^-}=\infty$ (Remember: $\frac{-}{-}=+$)

So, now we know that the graph approaches $\infty$ from the left as $x \to -2$, and $-\infty$ from the right as $x \to -2$.

*Consider the "end behaviour" (as $x \to \infty$ and $x \to -\infty$). This creates horizontal asymptotes.*

* As $x \to \infty$ then the numerator & denominator are practically the same value so $f(x) \to 1$. For large positive values of $x$, the numerator is always smaller than the denominator though, so it approaches 1 from below.
* Similarly, for $x \to -\infty$ $f(x) \to 1$, but this time it approaches 1 from above.

*Look at the sign of the function and note where its sign changes.*

For a rational function, the sign is positive when the sign of the numerator and demonitor are the same and negative when they're different.

Numerator: $x+1$ is positive when $x \gt -1$ and negative when $x \lt -1$
Denominator: $x+2$ is positive when $x \gt -2$ and negative when $x \lt -2$

The signs are only different when $-2 \lt x \lt -1$, so the function is negative between these values and positive everywhere else.

*Using information from the first derivative determine where the function is increasing/decreasing. Places where the first derivative is 0 give maximum or minimum values of the function.*

$f'(x)=\frac{(x+2)-(x+1)}{(x+2)^2}=\frac{1}{(x+2)^2}$

The first derivative is always positive for any value of $x$, (ignoring that vertical asymptote when $x=-2$). The square of a number must be positive, so the numerator and denominator are both positive, therefore the result must always be positive.

This means the function is always increasing, (expect for that vertical asymptote at $x=-2$)!

---

*Use information from the second derivative to determine the shape of the curve between the critical points. Also, consider where the second derivative is 0 to determine where the shape changes.*

$f'(x)=\frac{1}{(x+2)^2}=(x+2)^{-2}$

$f''(x)=-2(x+2)^{-3}=\frac{-2}{(x+2)^3}$

The numerator is always *negative*, so the results will be:

* *positive* when the denominator is *negative*, i.e., when $x+2<0$, i.e., $x<-2$. So the graph will be "concave up".
* *negative* when the denominator is *positive*, i.e., when $x+2>0$, i.e., $x>-2$. So the graph will be "concave down".

The second derivative is never 0. The point where it changes from "concave up" to "concave down" happens at that undefined point when $x=-2$.
