Consider a ring on a string held **fixed** at two ends. The ring is free to slide to any point. Find the position that the ring slides to.

The easiest way to think about this problem is to visualise it on a 2D plain. We can make one of the fixed ends the origin, $(0,0)$ and call the other $(a,b)$. (Remember that $a$ and $b$ are constants though because the ends are fixed.) Then call the position of the ring $(x,y)$. ($x$ and $y$ are variable.)

![](img\optimisation-ring-string-1.png)

Note that the ring will slide to the lowest point so it is a minimisation problem where we want to minimise $y$.

The constraint is that the length of the string (call it $L$) is constant. We can derive a formula for the length of the string from our diagram using the sum of two lengths calculated using the pythagorean theorem...

![](img\optimisation-ring-string-2.png)

$L=\sqrt{x^2+y^2}+\sqrt{(a-x)^2+(b-y)^2}$

We can differentiate this *wrt* $x$ remembering that $y$ is implicitly a function of $x$. It's quite involved though because of the square roots!

$L=\sqrt{x^2+y^2}+\sqrt{(a-x)^2+(b-y)^2}$

$L=(x^2+y^2)^\frac{1}{2}+((a-x)^2+(b-y)^2)^\frac{1}{2}$

$0=\frac{1}{2}(x^2+y^2)^{-\frac{1}{2}}(2x+2yy')+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}(2(a-x)(-1)+2(b-y)(-y'))$

$0=\frac{1}{2}(x^2+y^2)^{-\frac{1}{2}}2(x+yy')+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}(2(a-x)(-1)+2(b-y)(-y'))$

$0=(x^2+y^2)^{-\frac{1}{2}}(x+yy')+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}(2(a-x)(-1)+2(b-y)(-y'))$

$0=\frac{x+yy'}{(x^2+y^2)^\frac{1}{2}}+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}(2(a-x)(-1)+2(b-y)(-y'))$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}(2(a-x)(-1)+2(b-y)(-y'))$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+\frac{1}{2}((a-x)^2+(b-y)^2)^{-\frac{1}{2}}2((x-a)-y'(b-y))$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+((a-x)^2+(b-y)^2)^{-\frac{1}{2}}((x-a)-y'(b-y))$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+\frac{(x-a)-y'(b-y)}{((a-x)^2+(b-y)^2)^\frac{1}{2}}$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+\frac{(x-a)-y'(b-y)}{\sqrt{(a-x)^2+(b-y)^2}}$

$0=\frac{x+yy'}{\sqrt{x^2+y^2}}+\frac{x-a+y'(y-b)}{\sqrt{(a-x)^2+(b-y)^2}}$

The minimum is the critical point when $y'=0$. We could try to re-arrange to find $y'$ and then solve for $y'=0$, but that's quite a tricky re-arrangement. If we just stick $y'=0$ straight in then things simplify a bit...

$0=\frac{x+y0}{\sqrt{x^2+y^2}}+\frac{x-a+0(y-b)}{\sqrt{(a-x)^2+(b-y)^2}}$

$0=\frac{x}{\sqrt{x^2+y^2}}+\frac{x-a}{\sqrt{(a-x)^2+(b-y)^2}}$

So, by re-arranging so we get one term on each side, we get an equality that must be true when $y$ is at its mimumum...

$\frac{x}{\sqrt{x^2+y^2}}=\frac{-(x-a)}{\sqrt{(a-x)^2+(b-y)^2}}$

$\frac{x}{\sqrt{x^2+y^2}}=\frac{a-x}{\sqrt{(a-x)^2+(b-y)^2}}$

...which probably looks like it doesn't help us much at first glance, but if you look back at our sketch you can see that one term is the ratio between a side and the hypotenuse of the right-angle triangle on the left and the other term is the ratio between a side an the hypotenuse of the right-angle triangle on the right. In other words, they can both be expressed as the $\sin$ of the angles $\alpha$ and $\beta$ in the sketch below:

![](img\optimisation-ring-string-3.png)

$\sin\alpha=\frac{x}{\sqrt{x^2+y^2}}$

$\sin\beta=\frac{a-x}{\sqrt{(a-x)^2+(b-y)^2}}$

If we substitute those into our equation then we get something much more interesting...

$\sin\alpha=\sin\beta$

Or simply, $\alpha=\beta$.

This tells us that the ring will slide to the position where $\alpha=\beta$, but it doesn't tell us how to calculate that position. If the angles are equal then we can say that...

$\cos\alpha=\frac{-y}{\sqrt{x^2+y^2}}$, or, $-y=(\sqrt{x^2+y^2})\cos\alpha$

$\cos\alpha=\frac{b-y}{\sqrt{(a-x)^2+(b-y)^2}}$, or, $b-y=(\sqrt{(a-x)^2+(b-y)^2})\cos\alpha$

Adding those two equations you get...

$-y+b-y=(\sqrt{x^2+y^2})\cos\alpha+(\sqrt{(a-x)^2+(b-y)^2})\cos\alpha$

$b-2y=\cos\alpha(\sqrt{x^2+y^2}+\sqrt{(a-x)^2+(b-y)^2})$

Notice that we've already defined that term in brackets. It's the length of the string, $L$.

$b-2y=L\cos\alpha$

$-2y=L\cos\alpha-b$

$-y=\frac{1}{2}(L\cos\alpha-b)$

$y=\frac{1}{2}(b-L\cos\alpha)$

...but we want the answer in terms of $a$, $b$ and $L$ as those are the only knowns at the start of the problem. We can use the trigonometric identity:

$\sin^2\alpha+\cos^2\alpha=1$

$\cos^2\alpha=1-\sin^2\alpha$

$\cos\alpha=\sqrt{1-\sin^2\alpha}$

$L\cos\alpha=L\sqrt{1-\sin^2\alpha}$

...which doesn't look like it helps a lot because we've still got the unknown angle $\alpha$ in the equation. But we can get $\sin\alpha$ in terms of knowns by doing something similar with the formulae we have for $\sin\alpha$...

$\frac{x}{\sqrt{x^2+y^2}}=\sin\alpha$, or, $x=(\sqrt{x^2+y^2})\sin\alpha$

$\frac{a-x}{\sqrt{(a-x)^2+(b-y)^2}}=\sin\alpha$, or, $a-x=(\sqrt{(a-x)^2+(b-y)^2})\sin\alpha$

Adding those two equations you get...

$x+a-x=(\sqrt{x^2+y^2})\sin\alpha+(\sqrt{(a-x)^2+(b-y)^2})\sin\alpha$

$a=\sin\alpha(\sqrt{x^2+y^2}+\sqrt{(a-x)^2+(b-y)^2})$

$a=L\sin\alpha$

$\sin\alpha=\frac{a}{L}$

So substituting that in to our equation for $L\cos\alpha$...

$L\cos\alpha=L\sqrt{1-\sin^2\alpha}$

$=L\sqrt{1-(\frac{a}{L})^2}$

$=\sqrt{L^2-L^2(\frac{a}{L})^2}$

$=\sqrt{L^2-a^2}$

So:

$y=\frac{1}{2}(b-L\cos\alpha)$

$=\frac{1}{2}(b-\sqrt{L^2-a^2})$

To find $x$ we can use similar triangles as $\alpha=\beta$...

$\frac{x}{-y}=\frac{a-x}{b-y}$

$x(b-y)=(-y)(a-x)$

$x(b-y)=-ya+yx$

$x(b-y)-yx=-ya$

$x((b-y)-y)=-ya$

$x(b-2y)=-ya$

$x=\frac{-ya}{b-2y}$

...which is fine as long as we calculate $y$ first, but we might want to substitute to get a formulae for $x$ using only knowns:

$x=\frac{-ya}{b-2y}$

$x=\frac{-(\frac{1}{2}(b-\sqrt{L^2-a^2}))a}{b-2(\frac{1}{2}(b-\sqrt{L^2-a^2}))}$

$=\frac{-(\frac{1}{2}(b-\sqrt{L^2-a^2}))a}{b-(b-\sqrt{L^2-a^2})}$

$=\frac{-(\frac{1}{2}(b-\sqrt{L^2-a^2}))a}{b-b+\sqrt{L^2-a^2}}$

$=\frac{-(\frac{1}{2}(b-\sqrt{L^2-a^2}))a}{\sqrt{L^2-a^2}}$

$=-\frac{a}{2}\frac{b-\sqrt{L^2-a^2}}{\sqrt{L^2-a^2}}$

$=-\frac{a}{2}(\frac{b}{\sqrt{L^2-a^2}}-\frac{\sqrt{L^2-a^2}}{\sqrt{L^2-a^2}})$

$=-\frac{a}{2}(\frac{b}{\sqrt{L^2-a^2}}-1)$

$=\frac{a}{2}\big(1-\frac{b}{\sqrt{L^2-a^2}}\big)$

So, to summarise, the point that the ring slides to can be calculated from:

$x=\frac{a}{2}\big(1-\frac{b}{\sqrt{L^2-a^2}}\big)$

$y=\frac{1}{2}(b-\sqrt{L^2-a^2})$
