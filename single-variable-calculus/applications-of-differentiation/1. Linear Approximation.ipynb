If you have a curve, $y=f(x)$, it's approximately the same as its tangent line at a point, $x_0$, ***for values of $x$ close to $x_0$***.

![](img\linear-approximation.png)

The general formula for a tangent line is:

$y=mx+c$

* $m$ is the slope, i.e., the value of the derivative at $x_0$, ($f'(x_0)$).
* $c$ is the y-intercept - well, we'd normally say the value where the line crosses the vertical axis, but for linear approximation it's easier to define the line relative to the point, $x_0$, (as if the axes were at the black dotted lines shown above), so $c=f(x_0)$.
* This also means that when we're using the formula to read off values along the tangent line we need to offset $x$ relative to $x_0$, $(x-x_0)$.

The formula to use for linear approximation, then, is:

$y\approx f'(x_0)(x-x_0)+f(x_0)$

For example, if $y=\sin(x)$ then a linear approximation at $x_0=0$ is:

$y\approx f'(x_0)(x-x_0)+f(x_0)$

$\approx f'(0)(x-0)+f(0)$

$\approx f'(0)x+f(0)$

Now, $f(0)=0$ (because $\sin(0)=0$) and $f'(0)=1$ (because $\frac{d}{dx}\sin(x)=\cos(x)$ and $\cos(0)=1$). So...

$y\approx f'(0)x+f(0)$

$\approx 1x+0$

$\approx x$

```octave
x=[-2*pi:0.01:2*pi];
y=sin(x);

t=x;

plot(x,y, 'LineWidth', 2, x,t);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.5 6.5]);
set(gca, "ylim", [-3 3]);
set(gca, "box", "off");
```

Of course, for values further away from $x=0$ the linear approximation returns values nothing like the values of $\sin$, but if you zoom in around $x=0$ you can see that it's a good approximation for those values...

```octave
x=[-1:0.01:1];
y=sin(x);

t=x;

plot(x,y, 'LineWidth', 2, x,t);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

If you're still not convinced consider some example values...

$sin(0.01)$ using the approximation $y\approx x$ gives a value of $0.01$ where a calculator gives $0.00999...$

$sin(0.1)$ using the approximation $y\approx x$ gives a value of $0.1$ where a calculator gives $0.09983...$

$sin(1)$ using the approximation $y\approx x$ gives a value of $1$ where a calculator gives $0.84147...$

### Linear Approximation and the Definition of the Derivative

Another way of understanding this formula is by considering the definition of the derivative we started off with...

$f'(x_0)=\lim\limits_{\Delta x \to 0}\frac{\Delta f}{\Delta x}$

...which can be interpreted to mean:

$f'(x_0) \approx \frac{\Delta f}{\Delta x}$, when $\Delta x \approx 0$.

In other words, the instantaneous rate of change at a point, $x_0$, is nearly the same as the average rate of change of the function, $\frac{\Delta f}{\Delta x}$ over a small change in $x$, so we can use the former to approximate other values of the function close to that point.

We can also show that this leads to the same formula:

$f'(x_0) \approx \frac{\Delta f}{\Delta x}$

$f'(x_0)\Delta x \approx \Delta f$

$f'(x_0)(x-x_0) \approx f(x)-f(x_0)$ (because $\Delta x=x-x_0$, $\Delta f=f(x)-f(x_0)$)

$f'(x_0)(x-x_0)+f(x_0) \approx f(x)$

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)$

### Another worked example: ln(x) at x_0=1

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)$

where: $f(x)=\ln(x)$ and $x_0=1$

* $\ln(1)=0$ (because $e^0=1)$, so $f(x_0)=0$
* $\frac{d}{dx}\ln(x)=\frac{1}{x}$, so $f'(x_0)=\frac{1}{1}=1$

$f(x) \approx 1(x-1)+0$

$\approx (x-1)$

```octave
x=[0:0.01:2];
y=log(x);

t=x-1;

plot(x,y,'LineWidth',2,x,t);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

```octave
x=[0.8:0.01:1.2];
y=log(x);

t=x-1;

plot(x,y,'LineWidth',2,x,t);
set(gca, "xaxislocation", "zero");
set(gca, "box", "off");
```

### Some quick examples

#### cos at x_0=0

* $f(x)=\cos x$
* $f'(x)=-\sin x$
* $f(0)=1$
* $f'(0)=0$

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)\approx 0(x-0)+1\approx 1$

#### e^x at x_0=0

* $f(x)=e^x$
* $f'(x)=e^x$
* $f(0)=1$
* $f'(0)=1$

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)\approx 1(x-0)+1\approx x+1$

### Two slightly more challenging examples

#### ln(1+x) at x_0=0

* $f(x)=\ln(1+x)$
* $f'(x)=\frac{1}{1+x}$
* $f(0)=0$
* $f'(0)=1$

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)\approx 1(x-0)+0\approx x$

#### (1+x)^r at x_0=0

* $f(x)=(1+x)^r$
* $f'(x)=r(1+x)^{r-1}$
* $f(0)=1$
* $f'(0)=r$

$f(x) \approx f'(x_0)(x-x_0)+f(x_0)\approx r(x-0)+1\approx rx+1$
