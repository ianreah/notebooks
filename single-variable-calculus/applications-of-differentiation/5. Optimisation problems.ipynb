Optimisation problems often come down to finding the maximum or minimum point in a function. We can use some of the same techiques we used for sketching graphs to do this.

**Note:** we sometimes have to be a bit careful with the wording in these types of problems to avoid ambiguity between a ***point*** in a function (input & output pair), a ***value*** of a function (the output given a particular input), and the place ***where*** the value occurs (the input that produces a particular value).

To find the maximum &/or minimum point in a function we need to look at:

* critical points (i.e., points where the first derivative is zero) - these are local maxima or minima.
* end points
* points of discontinuity

# An example

Consider a wire of length 1, cut into two pieces. Bend each piece into a square. We want to figure out where to cut the wire in order to enclose as much area in the two squares as possible.

## Step 1: visualise the problem

![](img\optimisation-example-1.png)

$x$ defines the point where we cut the wire so that we have two lengths of $x$ and $1-x$. The sides of the two squares we can make from these lengths of wire are $\frac{x}{4}$ and $\frac{1-x}{4}$

## Step 2: define the formula we need to optimise

We're looking for the **maximum** combined area of the two squares, so we need to find the **maximum value** of the function...

$A = (\frac{x}{4})^2+(\frac{1-x}{4})^2$

$= \frac{x^2}{16}+\frac{(1-x)^2}{16}$

$= \frac{x^2+(1-x)^2}{16}$

$= \frac{1}{16}(2x^2-2x+1)$

## Step 3: find the critical points

*...i.e., the points where $A'$ is 0.*

$A'= \frac{1}{16}(4x-2)$

$A'= \frac{2x-1}{8}$

...which is 0 when $2x-1$ is zero...

$2x-1=0$

$2x=1$

$x=\frac{1}{2}$

So there's only one critical point: where $x=5$, the value of $A$ it at that point is:

$A = \frac{1}{16}(2x^2-2x+1)$

$= \frac{1}{16}(2(\frac{1}{2})^2-2(\frac{1}{2})+1)$

$= \frac{1}{32}$

## Step 4: consider the end behaviour

When $x=0$:

$A=\frac{1}{16}(2x^2-2x+1)$

$=\frac{1}{16}(2(0)^2-2(0)+1)$

$=\frac{1}{16}$

When $x=1$:

$A=\frac{1}{16}(2x^2-2x+1)$

$=\frac{1}{16}(2(1)^2-2(1)+1)$

$=\frac{1}{16}$

## Step 5: consider any points of discontinuity

There aren't any with this function.

## Step 6: answer the question

*...using the values calculated from steps 3, 4 & 5*

From the above we can say that the maximum **value** of this function is $\frac{1}{16}$ which occurs when $x=0$ or $x=1$. So the greatest area enclosed is $\frac{1}{16}$ which occurs when there is only one square. This might be considered "cheating" giving the question asked what's the greatest area when the wire is cut into 2. We should really say that $\frac{1}{16}$ is the limit when $x \to 0$ or $x \to 1$ (i.e., when the wire is cut close to either end).
