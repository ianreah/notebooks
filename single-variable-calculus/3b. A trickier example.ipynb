**Calculate the derivative with respect to $x$ of the following:**

$f(x)=\arctan\Big(\frac{x}{\sqrt{1-x^2}}\Big)$

To simplify things, let's just focus on the bit inside the brackets - call that $g(x)$...

$g(x)=\frac{x}{\sqrt{1-x^2}}$

...so...

$f(x)=\arctan(g(x))$

Now, let's find $g'(x)$:

$g(x)=\frac{x}{\sqrt{1-x^2}}$

$=\frac{x}{(1-x^2)^{\frac{1}{2}}}$

$=x(1-x^2)^{-\frac{1}{2}}$

...which we can differentiate with a combination of the chain rule & product rule...

$g'(x)=1.(1-x^2)^{-\frac{1}{2}}+x.(-\frac{1}{2}(1-x^2)^{-\frac{3}{2}}.(-2x))$

$=(1-x^2)^{-\frac{1}{2}}+x^2(1-x^2)^{-\frac{3}{2}}$

$=\frac{1}{(1-x^2)^{\frac{1}{2}}}+\frac{x^2}{(1-x^2)^{\frac{3}{2}}}$

$=\frac{(1-x^2)+x^2}{(1-x^2)^{\frac{3}{2}}}$

$g'(x)=\frac{1}{(1-x^2)^{\frac{3}{2}}}$

Now, treat it like an "Implicit differentiation & inverse function" problem:

$f(x)=\arctan(g(x))$

$\tan(f(x))=g(x)$

Differentiating both sides...

$\sec^2(f(x))f'(x)=g'(x)$

$\sec^2(f(x))f'(x)=\frac{1}{(1-x^2)^{\frac{3}{2}}}$

$f'(x)=\frac{1}{(1-x^2)^{\frac{3}{2}}\sec^2(f(x))}$

$=\frac{1}{(1-x^2)^{\frac{3}{2}}\sec^2\big(\arctan(\frac{x}{\sqrt{1-x^2}})\big)}$

...which is a very complicated version of the correct answer! Let's simplify it with some trigonometry...

Our initial function, $f(x)=\arctan\Big(\frac{x}{\sqrt{1-x^2}}\Big)$, can be interpreted as "the angle, say $\theta$, such that $\tan\theta=\frac{x}{\sqrt{1-x^2}}$, and if we look at [$\tan$ and $\sec$ in the context of the unit circle](0c.%20Trigonometric%20functions%20and%20the%20unit%20circle.ipynb#tan-and-sec), we can calculate $\sec^2$ of this angle using pythagoras...

![](img\sec-pythagoras.jpg)

$\sec^2=\Big(\frac{x}{\sqrt{1-x^2}}\Big)^2+1^2$

$=\frac{x^2}{1-x^2}+1$

$=\frac{x^2}{1-x^2}+\frac{1-x^2}{1-x^2}$

$=\frac{x^2+1-x^2}{1-x^2}$

$=\frac{1}{1-x^2}$

So, plugging that in to our derivative...

$f'(x)=\frac{1}{(1-x^2)^{\frac{3}{2}}\sec^2(f(x))}$

$=\frac{1}{(1-x^2)^{\frac{3}{2}}\frac{1}{1-x^2}}$

$=\frac{1-x^2}{(1-x^2)^{\frac{3}{2}}}$

That denominator, $(1-x^2)^{\frac{3}{2}}$, can be written as $(1-x^2)(1-x^2)^{\frac{1}{2}}$, using the [rules of exponents](0.%20References.ipynb#Exponents-refresher) ($a^1=a$ and $a^xa^y=a^{x+y}$), so...

$f'(x)=\frac{1-x^2}{(1-x^2)(1-x^2)^{\frac{1}{2}}}$

$=\frac{1}{(1-x^2)^{\frac{1}{2}}}$

$=\frac{1}{\sqrt{1-x^2}}$

...which is a much, much nicer version of the answer!
