For a function to be differentiable it [must first be continuous](1a. Differentiable functions must be continuous.ipynb), and the tangent at any point must be unamiguous (e.g., in a piecewise function the tangents where the pieces join must be equal).

-----


###Example

For what values of `a` and `b` is the following function...

1. continuous?
1. differentiable?

$f(x)=\begin{cases} 
      x^2+1 & x \gt 1 \\
      ax+b & x \leq 1 
   \end{cases}$
   
So, we know what the function looks like for $x \gt 1$...

```octave
x = [1:0.1:4];
y1 = x.^2 + 1;

plot(x,y1,'LineWidth',2);
set(gca, "xlim", [-2 4]);
set(gca, "ylim", [-5 15]);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

...and for $x \leq 1$ the function is linear.

**Consider the 3 conditions for continuity**

In our example, $x_0=1$.

Condition 1 is satisfied because the value at $x=1$ is explicitly included in the range of the second piece of the function, $f(1)=ax+b$.

To satisfy condition 2 we need values for `a` and `b` such that $\lim\limits_{x \to 1^+} f(x) = \lim\limits_{x \to 1^-} f(x)$...

$\lim\limits_{x \to 1^+} f(x) = \lim\limits_{x \to 1^+} (x^2+1)=2$

$\lim\limits_{x \to 1^-} f(x) = \lim\limits_{x \to 1^-} (ax+b) = a + b$

$a + b$ must equal 2 for the function to be continuous.

Finally, to satisfy condition 3, the above limits and the defined value for $x=1$ must all be equal:

$2 = a + b = f(1) = a.1 + b$

So, the function is continuous for any values of `a` and `b` where $a+b=2$.

**Differentiable?**

For the piecewise function to be differentiable it must first be continuous (which we've just proven it is if $a+b=2$), **and** the tangents of the pieces where they join must be equal.

To calculate the tangents we use differentiation:

$\frac{d}{dx}(x^2+1)=2x$, which equals 2 when $x=1$.

$\frac{d}{dx}(ax+b)=a$, which must equal 2 when $x=1$ for the tangents to be equal and the function to be differentiable. In other words $a=2$.

So, there is only a single pair of values for `a` and `b` for the function to be differentiable: $a=2$, and $a+b=2$ (or $b=0$).



```octave
a = 2;
b = 0;

x = [-2:0.1:4];
y1 = x.^2 + 1;
y2 = a*x + b;

y = (y1.*(x>1))+(y2.*(x<=1));

plot(x,y,'LineWidth',2);
set(gca, "xlim", [-2 4]);
set(gca, "ylim", [-5 15]);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```
