Let $h(x)=\sin(x)+\sqrt{3}\cos(x)$

For which values of $x$ does $h'(x)=0$?

---

First lets plot $h(x)$...

```octave
x=[-5:0.01:5];
y=sin(x)+(sqrt(3).*cos(x));

plot(x,y,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

Now, to calculate the derivative, which [rules](2. Using the definition to derive some differentiation rules.ipynb) do we need?


First, the [sum rule](2. Using the definition to derive some differentiation rules.ipynb#Sum-rule) for the addition, (which says that the derivative of a sum is equal to the sum of the derivatives)...

$h'(x)=\frac{d}{dx}(\sin x)+\frac{d}{dx}(\sqrt{3}\cos x)$

Then applying the [constant multiple rule](2. Using the definition to derive some differentiation rules.ipynb#Constant-multiple-rule) to the $\sqrt{3} \times ...$ part gives us...

$h'(x)=\frac{d}{dx}(\sin x)+\sqrt{3}\frac{d}{dx}(\cos x)$

Finally, of course, the trigonometry rules for [sin](2. Using the definition to derive some differentiation rules.ipynb#sin) and [cos](2. Using the definition to derive some differentiation rules.ipynb#cos) give us...

$h'(x)=\cos x+\sqrt{3}(-\sin x)$

$h'(x)=\cos(x)-\sqrt{3}\sin(x)$

Now, let's plot that...

```octave
x=[-5:0.01:5];
y=cos(x)-(sqrt(3).*sin(x));

plot(x,y,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

But, for which values of $x$ is the derivative equal to $0$? You can read it off the graph and guess the values, but a better way is to solve the equation...

$\cos(x)-\sqrt{3}\sin(x)=0$

(You can always come back and make sure the result matches what you see in the above graph.)

$\cos(x)-\sqrt{3}\sin(x)=0$

$\cos(x)=\sqrt{3}\sin(x)$

$\frac{\cos(x)}{\cos(x)}=\frac{\sqrt{3}\sin(x)}{\cos(x)}$

$1=\sqrt{3}\frac{\sin(x)}{\cos(x)}$

We know from the [trigonometric identities](0. References.ipynb#Trigonometric-identities) that...

$\frac{sin(x)}{cos(x)}=\tan(x)$

Therefore...

$1=\sqrt{3}\tan(x)$

$\frac{1}{\sqrt{3}}=\tan(x)$

$\tan(x)=\frac{1}{\sqrt{3}}$

$x=\tan^{-1}(\frac{1}{\sqrt{3}})=\frac{\pi}{6}$ radians

Of course, $\tan$ is a periodic function which is $0$ every $\pi$ radians, so there are actually infinite values of $x$ where the derivative is equal to $0$...

$x= ..., \frac{-5\pi}{6}, \frac{\pi}{6}, \frac{7\pi}{6}, \frac{13\pi}{6}, ...$
