Assume all letters represent constants, except for the independent and dependent variables occurring in the derivative.

$m=\frac{m_0}{\sqrt{1-v^2/c^2}}$, $\frac{dm}{dv}=?$

There might be a simpler, cleverer way to do it but the following steps make sense to me...

### Step 1: Constant multiple rule
...to get $m_0$ out of the derivative.

$m=\frac{m_0}{\sqrt{1-v^2/c^2}}$

$m=m_0\frac{1}{\sqrt{1-v^2/c^2}}$

$\frac{dm}{dv}=m_0\frac{d}{dv}\frac{1}{\sqrt{1-v^2/c^2}}$

### Step 2: Consider the term containing the dependant variable in isolation

$f(v)=1-v^2/c^2$

$f'(v)=-\frac{2v}{c^2}$

### Step 3: Chain rule to get the derivative of the square root

Suppose we've got a function for taking the square root:

$g(x)=\sqrt{x}$

...which can be written as $g(x)=x^{\frac{1}{2}}$ so it can be differentiated using the power rule:

$g'(x)=\frac{1}{2}x^{-\frac{1}{2}}$

The denominator in the original equation can be thought of as $g(f(v))$, which can be differentiated using the chain rule...

$\frac{d}{dv}g(f(v))=g'(f(v)).f'(v)$

$=\frac{1}{2}(1-v^2/c^2)^{-\frac{1}{2}}(-\frac{2v}{c^2})$

$=-\frac{2v}{2c^2\sqrt{1-v^2/c^2}}$

### Step 4: Quotient rule
...to get $\frac{d}{dv}\frac{1}{\sqrt{1-v^2/c^2}}$

Using the functions we've already introduced...

$\frac{d}{dv}\frac{1}{\sqrt{1-v^2/c^2}}=\frac{d}{dv}\frac{1}{g(f(v))}$

So, using the quotient rule...

$\frac{d}{dv}\frac{1}{g(f(v))}=\frac{0.(g(f(v)))-1.(g'(f(v)))}{(g(f(v)))^2}$

$=-\frac{g'(f(v))}{(g(f(v)))^2}$

We've already calculated the numerator, and, because $g$ is the square root function then its square is just the original value...

$\frac{d}{dv}\frac{1}{g(f(v))}=-(-\frac{2v}{2c^2\sqrt{1-v^2/c^2}})/(1-v^2/c^2)$

$=\frac{2v}{2c^2\sqrt{1-v^2/c^2}}\times\frac{1}{1-v^2/c^2}$

$=\frac{2v}{2c^2(1-v^2/c^2)^\frac{1}{2}}\times\frac{1}{1-v^2/c^2}$

$=\frac{v}{c^2(1-v^2/c^2)^\frac{3}{2}}$

### And finally...

$\frac{dm}{dv}=m_0\frac{d}{dv}\frac{1}{\sqrt{1-v^2/c^2}}$

$=m_0\frac{v}{c^2(1-v^2/c^2)^\frac{3}{2}}$

$=\frac{m_0v}{c^2(1-v^2/c^2)^\frac{3}{2}}$

### Now try...

$F=\frac{mg}{(1+r^2)^{\frac{3}{2}}}$, $\frac{dF}{dr}=?$

Start with the denominator by itself...

$f(r)=(1+r^2)^{\frac{3}{2}}$

Differentiate using the chain rule:

$f'(r)=\frac{3}{2}(1+r^2)^{\frac{1}{2}}2r$

$=\frac{6r}{2}(1+r^2)^{\frac{1}{2}}$

$=3r(1+r^2)^{\frac{1}{2}}$

Now use the quotient rule to get the whole derivative:

$\frac{dF}{dr}=\frac{0.(1+r^2)^{\frac{3}{2}}-mg.3r(1+r^2)^{\frac{1}{2}}}{(1+r^2)^{\frac{3}{2}}(1+r^2)^{\frac{3}{2}}}$

$=-\frac{mg.3r(1+r^2)^{\frac{1}{2}}}{(1+r^2)^{\frac{3}{2}}(1+r^2)^{\frac{3}{2}}}$

$=-\frac{3mgr}{(1+r^2)^{\frac{5}{2}}}$
