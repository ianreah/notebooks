## sin and cos

The $\cos$ and $\sin$ functions can be thought of as the $x$ and $y$ co-ordinates, respectively, of a point, $P$, moving around the unit circle.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8b/Webysther_20160305_-_Ciclo_trigonom%C3%A9trico.svg/505px-Webysther_20160305_-_Ciclo_trigonom%C3%A9trico.svg.png)

*By Webysther (Own work) [<a href="http://creativecommons.org/licenses/by-sa/4.0">CC BY-SA 4.0</a>], <a href="https://commons.wikimedia.org/wiki/File%3AWebysther_20160305_-_Ciclo_trigonom%C3%A9trico.svg">via Wikimedia Commons</a>*

With the angle $\theta$ expressed in radians you can see that both functions are periodic with period $2\pi$.

$\cos$ going $1\to0$ over the first quadrant ($\theta=0\to\frac{\pi}{2}$), $0\to-1$ over the second quadrant ($\theta=\frac{\pi}{2}\to\pi$), $-1\to0$ over the third quadrant ($\theta=\pi\to\frac{3\pi}{2}$), and $0\to1$ over the fourth quadragnt ($\theta=\frac{3\pi}{2}\to2\pi$). Similarly, it's easy to see that $\sin$ goes $0\to1$, $1\to0$, $0\to-1$, and, $-1\to0$ over the four quadrants respectively.

```octave
theta=[-2*pi:0.01:2*pi];
x=cos(theta);
y=sin(theta);

plot(theta,x,'LineWidth',3,theta,y,'LineWidth',3);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.7 6.7]);
set(gca, "ylim", [-1.35 1.35]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "box", "off");

legend ({'cos(\theta)','sin(\theta)'});
legend boxoff;
```

## tan and sec

There are a couple of ways to think about the $\tan$ function in terms of the unit circle. The one that seems to make most sense to me is by imagining a vertical line tangential to the unit circle. Then extend a line from the center of the circle, through the point on the circle ($P$, above) until it touches the vertical tangential line. The $\tan$ function gives the vertical height to this point...

![](https://upload.wikimedia.org/wikipedia/commons/thumb/4/45/Unitcircledefs.svg/328px-Unitcircledefs.svg.png)

*By en:User:Michael Hardy (original); Pbroks13 (talk) (redraw) (en:Image:Unitcircledefs.png) [Public domain], <a href="https://commons.wikimedia.org/wiki/File%3AUnitcircledefs.svg">via Wikimedia Commons</a>*

You can see that, like $\sin$, the value of this function increases as $\theta$ increases over the first quadrant ($0\to\frac{\pi}{2})$, but does very quickly start to increase much more rapidly than $\sin$, approaching $+\infty$ as $\theta$ approaches $\frac{pi}{2}$. Of course, when $\theta=\frac{pi}{2}$ the two lines are parallel and never touch. In fact, the value of $\tan$ is undefined when $\theta=\frac{pi}{2}$. It has a different limit approaching from each direction:

$\lim\limits_{\theta\to\frac{pi}{2}^+}(\tan\theta)=+\infty$

...but as soon as $\theta$ goes above $\frac{pi}{2}$ the line flips over and crosses the vertical tangential line and a point far below the point where the tangential line touches the circle...

$\lim\limits_{\theta\to\frac{pi}{2}^+}(\tan\theta)=-\infty$

This also means that over the second quadrant ($\frac{pi}{2}\to\pi$) the $\tan$ function increases $-\infty\to0$ (undefined when $\theta=\frac{pi}{2}$ don't forget). The pattern repeats over the third & fourth quadrants.


```octave
theta=[-2*pi:0.01:2*pi];
t=tan(theta);

mask = t < 6 & t > -6;
t=t.*mask;
t(t==0)=NaN;

plot(theta,t,'LineWidth',3, [-3*pi/2,-3*pi/2], [-5, 5],'k', [-pi/2,-pi/2], [-5, 5],'k', ...
                            [3*pi/2,3*pi/2], [-5, 5],'k', [pi/2,pi/2], [-5, 5],'k');
                            
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.7 6.7]);
set(gca, "ylim", [-5 5]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "box", "off");
```

It's easy to see how to calculate $\tan$ in terms of $\sin$ and $\cos$ by considering the two similar triangles: $\triangle{QOS}$ and $\triangle{POR}$...

![](img\tan.png)

Corresponding sides in the two triangles must be in the same proportion:

$\frac{QS}{PR}=\frac{OS}{OR}$

$\frac{\tan\theta}{\sin\theta}=\frac{1}{\cos\theta}$

$\tan\theta=\frac{\sin\theta}{\cos\theta}$

The $\sec$ is the length of the hypotenuse in the larger of the two triangles, $\triangle{QOS}$.

$\frac{OQ}{OP}=\frac{OS}{OR}$

$\frac{\sec\theta}{1}=\frac{1}{\cos\theta}$

$\sec\theta=\frac{1}{\cos\theta}$

And looks like this: (Notice it has the same asymptotes as $\tan$)

```octave
theta=[-2*pi:0.01:2*pi];
t=sec(theta);

mask = t < 6 & t > -6;
t=t.*mask;
t(t==0)=NaN;

plot(theta,t,'LineWidth',3, [-3*pi/2,-3*pi/2], [-5, 5],'k', [-pi/2,-pi/2], [-5, 5],'k', ...
                            [3*pi/2,3*pi/2], [-5, 5],'k', [pi/2,pi/2], [-5, 5],'k');
                            
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.7 6.7]);
set(gca, "ylim", [-5 5]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "box", "off");
```

## cot and csc

$\cot$ and $\csc$ are similar to $\tan$ and $\sec$, but relating to the triangle made with a horizontal tangential line...

![](https://upload.wikimedia.org/wikipedia/commons/thumb/2/2f/Unitcirclecodefs.svg/402px-Unitcirclecodefs.svg.png)

*By en:User:Michael Hardy (original); Pbroks13 (talk) (redraw) (en:Image:Unitcirclecodefs.png) [Public domain], <a href="https://commons.wikimedia.org/wiki/File%3AUnitcirclecodefs.svg">via Wikimedia Commons</a>*

Working from ratios of the sides in the similar triangles again, we get...

$\frac{\cot\theta}{\cos\theta}=\frac{1}{\sin\theta}$

$\cot\theta=\frac{\cos\theta}{\sin\theta}=\frac{1}{\tan\theta}$

$\frac{\csc\theta}{1}=\frac{1}{\sin\theta}$

$\csc\theta=\frac{1}{\sin\theta}$

```octave
theta=[-2*pi:0.01:2*pi];
t=cot(theta);
c=csc(theta);

mask = t < 6 & t > -6;
t=t.*mask;
t(t==0)=NaN;

mask = c < 6 & c > -6;
c=c.*mask;
c(c==0)=NaN;

plot(theta,t,'LineWidth',3, theta,c,'LineWidth',3,...
                            [-2*pi,-2*pi], [-5, 5],'k', [-pi,-pi], [-5, 5],'k', ...
                            [2*pi,2*pi], [-5, 5],'k', [pi,pi], [-5, 5],'k');
                            
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "xlim", [-6.7 6.7]);
set(gca, "ylim", [-5 5]);
set(gca,'xtick',[-2*pi:pi/2:2*pi])
set(gca,'xticklabel',{'-2pi','-3pi/2','-pi','-pi/2','0', 'pi/2', 'pi', '3pi/2', '2pi'})
set(gca, "box", "off");
```
