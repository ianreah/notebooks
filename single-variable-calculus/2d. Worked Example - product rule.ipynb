## Simple Example

$h(x)=x^2\sin(x)$

Think of it as the product of two functions: $f(x)=x^2$ and $g(x)=\sin(x)$.

```octave
x=[-20:0.01:20];
f=x.^2;
g=sin(x);
h=f.*g;

plot(x,h,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

$h'(x)=f'(x)g(x)+f(x)g'(x)$ *(product rule)*

$f'(x)=2x$ *(power rule)*

$g'(x)=cos(x)$ *(sin rule)*

$h'(x)=2x\sin(x)+x^2\cos(x)$

```octave
x=[-20:0.01:20];
dh=(2.*x.*sin(x))+(x.^2.*cos(x));

plot(x,dh,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

## Product rule with 3 functions

Suppose $f$, $g$, and $h$ are all functions of $x$ and we want to work out the derivative of the product of the three functions. We can start by introducing a fourth function, $k$, as the product of two of the functions:

$k(x)=g(x)h(x)$

We can see that:

$f(x)g(x)h(x)=f(x)k(x)$

So we can use the product rule to find the derivative of that:

$\frac{d}{dx}f(x)g(x)h(x)=f'(x)k(x)+f(x)k'(x)$

And, of course, we know how to find the derivative of $k$ using the product rule as well:

$k'(x)=g(x)'h(x)+g(x)h(x)'$

So, the derivative of the product of the three functions is therefore:

$\frac{d}{dx}f(x)g(x)h(x)=f'(x)g(x)h(x)+f(x)(g(x)'h(x)+g(x)h(x)')$

$=f'(x)g(x)h(x)+f(x)g(x)'h(x)+f(x)g(x)h(x)'$

The same approach can be used for the product of any number of functions.

**Example:**

$f(x)=x^2\sin(x)\cos(x)$

$f'(x)=\frac{d}{dx}(x^2)\sin(x)\cos(x)+x^2\frac{d}{dx}(\sin(x))\cos(x)+x^2\sin(x)\frac{d}{dx}(\cos(x))$

$f'(x)=2x\sin(x)\cos(x)+x^2\cos(x)\cos(x)-x^2\sin(x)\sin(x)$

$f'(x)=2x\sin(x)\cos(x)+x^2\cos^2(x)-x^2\sin^2(x)$
