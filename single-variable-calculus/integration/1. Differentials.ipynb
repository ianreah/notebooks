## A conceptual explanation of differentials

To move on from derivatives and start considering integrals we'll begin by introducing the term "differential". Now, trying to define exactly what a "differential" is can be fraught with danger ([see the first few paragraphs of this blogpost](http://math.blogoverflow.com/2014/11/03/more-than-infinitesimal-what-is-dx/)). [Some have argued](https://math.stackexchange.com/a/1358957) that it's less important to understand rigorously what a differential *is* as long as you know how to use them, etc. And they could well be right but I felt a little uncomfortable trying to follow some of the material without having a vague understanding of the concept. So (after a little bit of reading around) here's my attempt at explaining "just enough" of the concept...

Firstly, the notation of a differential, e.g., $dx$, comes from [Leibniz's notation for the derivative](../differentiation/0b.%20A%20bit%20about%20notation.ipynb#Leibniz's-notation):

$\frac{dy}{dx}$

The original intention was that this represented a ratio of infinitesimal values. The "infinitesimal change in $y$ produced by an infinitesmal change in $x$" divided by the "infinitesmal change in $x$". But this lead to a lot of problems (like the ones hinted to in that blog post linked to above - see also [this stackexchange answer](https://math.stackexchange.com/a/21209) for more about the problems of treating this as a ratio of infinitesimal values).

That's why (as [explained right at the start of these notes](../differentiation/1.%20Definition%20of%20Derivative.ipynb)) the derivative is now defined in terms of a limit:

$\lim\limits_{\Delta x\to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$

The notation $\frac{dy}{dx}$ doesn't actually represent a quotient but a single limit. So what do the differentials, $dx$ and $dy$, mean when they appear on their own. The concept I'm going with for now is that they represent changes in $x$ and $y$ along the tangent line to the curve at the point in question, rather than along the curve itself.

$\frac{dy}{dx}=f'(x)$

$dy=f'(x)dx$

And for small values they give a reasonable approximation to the curve itself - which brings us back to linear approximation...

## Differentials and Linear Approximation

**Example:** Approximate $(64.1)^{\frac{1}{3}}$

$y=x^{\frac{1}{3}}$

$dy=\frac{1}{3}x^{-\frac{2}{3}}dx$

When $x=64$...

$y=64^{\frac{1}{3}}=4$

$dy=\frac{1}{3}64^{-\frac{2}{3}}dx$

$=\frac{1}{3}{\frac{1}{16}}dx=\frac{1}{48}dx$

At $64.1$ $dx=\frac{1}{10}$, so...

$(64.1)^{\frac{1}{3}}\approx 4 + \frac{1}{48}\frac{1}{10}\approx 4\frac{1}{480}\approx 4.002$

If you were to do it with the linear approximation formula we've used before you'd get the same answer with essentially the same calculations. It's really only the notation that has changed...

$y \approx f(x_0)+f'(x_0)(x-x_0)$

$x_0=64$

$f(x_0)=64^{\frac{1}{3}}=4$

$f'(x_0)=\frac{1}{3}64^{-\frac{2}{3}}=\frac{1}{48}$

$x-x_0=64.1-64=\frac{1}{10}$

$y \approx f(x_0)+f'(x_0)(x-x_0)$

$\approx 4+\frac{1}{48}\frac{1}{10} \approx 4.002$

## Computing Differentials

If $y=7u^9+34-5u^{-3}$ how would we compute the differential $dy$? (Note: we can by-pass $y$ altogether and the differential could be written as $d(7u^9+34-5u^{-3})$)

The differential can be distributed using the same rules as derivatives...

$d(7u^9+34-5u^{-3})$

$=d(7u^9)+d(34)-d(5u^{-3})$ [[Sum Rule](../differentiation/2.%20Using%20the%20definition%20to%20derive%20some%20differentiation%20rules.ipynb#Sum-rule)]

$=7d(u^9)+d(34)-5d(u^{-3})$ [[Constant Multiple Rule](../differentiation/2.%20Using%20the%20definition%20to%20derive%20some%20differentiation%20rules.ipynb#Constant-multiple-rule)]

Then each part is calculated using the [Chain Rule](../differentiation/2.%20Using%20the%20definition%20to%20derive%20some%20differentiation%20rules.ipynb#Chain-rule)...

$=63u^8du+15u^{-4}du$

Another example:

$d(\sin\theta\cos\theta)$

...which can be solved with the [Product Rule](../differentiation/2.%20Using%20the%20definition%20to%20derive%20some%20differentiation%20rules.ipynb#Product-rule)...

$=d(\sin\theta)\cos\theta+\sin\theta d(\cos\theta)$

And, again, each part is calculated using the [Chain Rule](../differentiation/2.%20Using%20the%20definition%20to%20derive%20some%20differentiation%20rules.ipynb#Chain-rule)...

$=(\cos\theta d\theta)cos\theta+\sin\theta(-\sin\theta d\theta)$

$=\cos^2\theta d\theta-\sin^2\theta d\theta$

$=(\cos^2\theta-\sin^2\theta)d\theta$

$=(\cos 2\theta)d\theta$
