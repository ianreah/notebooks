So far, to differentiate a function with respect to some variable, say $x$, we've always written the function **explicitly** as a function of $x$. In other words, either something like $f(x)=3x^2-x+5$ or $y=3x^2-x+5$.

But suppose we're given an equation that does not directly describe a function of $x$. For example,

$x^2+y^2=1$

How could you differentiate this with respect to $x$? Well, you could rearrange it into an **explicit** function of $x$...

$x^2+y^2=1$

$y^2=1-x^2$

$y^2=1-x^2$

$y=\pm\sqrt{1-x^2}$

The problem is that this still isn't a function because there are two possibilities for the result - positive or negative. It's actually two functions stitched together...

```octave
x=[-1:0.01:1];
y_positive=sqrt(1-x.^2);
y_negative=-sqrt(1-x.^2);

plot(x,y_positive,'LineWidth',2,x,y_negative,'LineWidth',2);
set(gca, "xaxislocation", "zero");
set(gca, "yaxislocation", "zero");
set(gca, "box", "off");
```

...so to be thorough we should probably differentiate them each separately...

### Positive case:

$y=\sqrt{1-x^2}=(1-x^2)^{\frac{1}{2}}$

...which we can differentiate with the chain rule:

$y'=\frac{1}{2}(1-x^2)^{-\frac{1}{2}}(-2x)$

$=\frac{-2x}{2\sqrt{1-x^2}}$

$=\frac{-x}{\sqrt{1-x^2}}$

...and notice that the denominator is the same as the right hand side of the equation we started with, so we can substitute for $y$:

$y'=\frac{-x}{y}$

```octave
x=[-1:0.01:1];
y=sqrt(1-x.^2);
y_diff=-x./y;

plotyy(x,y,x,y_diff);
set(gca, "xaxislocation", "zero");
```


### Negative case:

$y=-\sqrt{1-x^2}=-(1-x^2)^{\frac{1}{2}}$

$y'=-\frac{1}{2}(1-x^2)^{-\frac{1}{2}}(-2x)$

$=\frac{2x}{2\sqrt{1-x^2}}$

$=\frac{x}{\sqrt{1-x^2}}$

...this time the denominator is the negative of the equation we started with:

$y'=\frac{x}{-y}$

```octave
x=[-1:0.01:1];
y=-sqrt(1-x.^2);
y_diff=x./-y;

plotyy(x,y,x,y_diff);
set(gca, "xaxislocation", "zero");
```

A simpler way to approach this is to imagine $y$ as an implicit function of $x$, say $f(x)$...

$x^2+(f(x))^2=1$

...and recognising that any terms involving $f$ should be differentiated using the chain rule:

$2x+2(f(x))f'(x)=0$

Of course, once you understand this you don't really need to introduce $f(x)$ at all:

$x^2+y^2=1$

$2x+2yy'=0$

And then rearrange to find $y'$:

$2x+2yy'=0$

$2yy'=-2x$

$y'=\frac{-2x}{2y}$

$y'=\frac{-x}{y}$

Of course you're still going to have to deal with the initial rearrangement at some point to calculate values for both $x$ and $y$, and (in this example) for every value of $x$ you'll have two values of $y$ (positive and negative), but whichever pair of values you plug in to the equation for $y'$ it will give you the gradient at that point. The point is that in many cases it's far easier to differentiate implicitly like this than to differentiate the explicit rearranged function (or functions).
