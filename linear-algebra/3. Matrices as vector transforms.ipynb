A set of simultaneous equations like...

$\begin{array}{lcl}
 2a & + & 3b & = & 8 \\
10a & + &  b & = & 13 \end{array}$

...can be written as a matrix multiplication:

$\begin{bmatrix}
2 & 3 \\
10 & 1
\end{bmatrix}
\begin{bmatrix}
a \\
b
\end{bmatrix}=
\begin{bmatrix}
8 \\
13
\end{bmatrix}$

(Expand out the matrix multiplication to convince yourself they mean the same thing.)

So the question posed by the initial simultaneous equations: "*what is $a$ and $b$?*"

...can be rephrased as: "*what vector, $\begin{bmatrix}
a \\
b
\end{bmatrix}$, can be multiplied by the matrix, $\begin{bmatrix}
2 & 3 \\
10 & 1
\end{bmatrix}$, to give the new vector, $\begin{bmatrix}
8 \\
13
\end{bmatrix}$?*"

The matrix can be thought of as a function that **transforms** a vector to give another vector.

Now, imagine multiplying the matrix with each of the unit basis vectors in a two-dimensional vector space, $\begin{bmatrix}
1 \\
0
\end{bmatrix}$ and $\begin{bmatrix}
0 \\
1
\end{bmatrix}$.

$\begin{bmatrix}
2 & 3 \\
10 & 1
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}=
\begin{bmatrix}
(2\times1) + (3\times0) \\
(10\times1) + (1\times0)
\end{bmatrix}
=
\begin{bmatrix}
2 \\
10
\end{bmatrix}$

$\begin{bmatrix}
2 & 3 \\
10 & 1
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}=
\begin{bmatrix}
(2\times0) + (3\times1) \\
(10\times0) + (1\times1)
\end{bmatrix}
=
\begin{bmatrix}
3 \\
1
\end{bmatrix}$

Notice how each column in the matrix operates on each axis in the vector space. The result of applying the transform to any vector in the space is then a vector sum of the transformed basis vectors. For example, if we apply the above transform to $\begin{bmatrix}
3 \\
2
\end{bmatrix}$

$\begin{bmatrix}
2 & 3 \\
10 & 1
\end{bmatrix}
\begin{bmatrix}
3 \\
2
\end{bmatrix}=
\begin{bmatrix}
(2\times3) + (3\times2) \\
(10\times3) + (1\times2)
\end{bmatrix}
=
\begin{bmatrix}
12 \\
32
\end{bmatrix}$

...can also be thought of in terms of the vector sum of the transformed basis vectors using the columns of the matrix...

$3 \times \begin{bmatrix}
2 \\
10
\end{bmatrix} +
2 \times \begin{bmatrix}
3 \\
1
\end{bmatrix}
=
\begin{bmatrix}
12 \\
32
\end{bmatrix}$

## Some typical transform matrices

Remember, it can be useful to think of the columns in the transform matrix as describing the positions of the unit basis vectors after the transform. You can also use [this interactive visualisation](https://shadanan.github.io/MatVis/) to experiment with the transforms.

![](img\matrix-transform-no-transform.png)

**Identity Matrix:** a matrix that doesn't change anything. 

$\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}$

**Stretching Matrix:** scales the unit basis factors.

$\begin{bmatrix}
\frac{1}{2} & 0 \\
0 & 2
\end{bmatrix}$

![](img\matrix-transform-stretching.png)

**Inversion Matrix:**

$\begin{bmatrix}
-1 & 0 \\
0 & -1
\end{bmatrix}$

![](img\matrix-transform-inversion.png)

**Mirroring Matrices:**

*$45^\circ$ mirror...*

$\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}$

![](img\matrix-transform-mirror.png)

*$-45^\circ$ mirror...*

$\begin{bmatrix}
0 & -1 \\
-1 & 0
\end{bmatrix}$

*Vertical mirror...*
(flips the x-axis)

$\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}$

*Horizontal mirror...*
(flips the y-axis)

$\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}$

**Shearing Matrices:**

$\begin{bmatrix}
1 & 1 \\
0 & 1
\end{bmatrix}$

![](img\matrix-transform-shearing.png)

**Rotating Matrices:**

*$90^\circ$ anticlockwise...*

$\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}$

![](img\matrix-transform-rotation.png)

*$\theta$...*

$\begin{bmatrix}
\cos\theta & \sin\theta \\
-\sin\theta & \cos\theta
\end{bmatrix}$
