Consider the following transform:

$T=
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}
$

...which is a shear with a vertical stretch transform:

![](img\shear-and-vertical-stretch.png)

Suppose we wanted to apply it successively to an initial vector:

$\vec v_0=
\begin{bmatrix}
-1 \\
1
\end{bmatrix}
$

First, let's work out the vectors after each application of the transform:

$\vec v_1=T\vec v_0=
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}
\begin{bmatrix}
-1 \\
1
\end{bmatrix}=
\begin{bmatrix}
-1+1 \\
0+2
\end{bmatrix}=
\begin{bmatrix}
0 \\
2
\end{bmatrix}$

$\vec v_2=T\vec v_1=
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}
\begin{bmatrix}
0 \\
2
\end{bmatrix}=
\begin{bmatrix}
0 + 2\\
0 + 4
\end{bmatrix}=
\begin{bmatrix}
2\\
4
\end{bmatrix}$

Or, if we don't care about $\vec v_1$, we could calculate $T^2$ and apply that to $\vec v_0$ to get $\vec v_2$:

$\vec v_2= TT\vec v_0=T^2\vec v_0$

$T^2=
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}=
\begin{bmatrix}
1+0 & 1+2 \\
0+0 & 0+4
\end{bmatrix}=
\begin{bmatrix}
1 & 3 \\
0 & 4
\end{bmatrix}$

$T^2\vec v_0=
\begin{bmatrix}
1 & 3 \\
0 & 4
\end{bmatrix}
\begin{bmatrix}
-1 \\
1
\end{bmatrix}=
\begin{bmatrix}
-1+3 \\
0+4
\end{bmatrix}=
\begin{bmatrix}
2 \\
4
\end{bmatrix}$

...which of course gives us the same answer.

Now let's see if we can make an eigenbasis for this transform...

$\lambda^2-(a+d)\lambda + ad-bc=0$

$\lambda^2-(1+2)\lambda + 1\times 2-1\times 0=0$

$\lambda^2-3\lambda + 2=0$

$(\lambda-1)(\lambda-2)=0$

...giving us eigenvalues of $\lambda=1$ and $\lambda=2$.

When $\lambda=1$:

$(T-\lambda I)\vec x=0$

$\bigg(
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}-
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
\bigg)\vec x=0$

$\begin{bmatrix}
0 & 1 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}=0$

$\begin{bmatrix}
0\times x_1 + 1\times x_2\\
0\times x_1 + 1\times x_2
\end{bmatrix}=0$

$\begin{bmatrix}
x_2\\
x_2
\end{bmatrix}=0$

So any vector where $x_2=0$ will be an eigenvector with an eigenvalue of $1$, e.g., $\begin{bmatrix}1\\0\end{bmatrix}$.

When $\lambda=2$:

$(T-\lambda I)\vec x=0$

$\bigg(
\begin{bmatrix}
1 & 1 \\
0 & 2
\end{bmatrix}-
\begin{bmatrix}
2 & 0 \\
0 & 2
\end{bmatrix}
\bigg)\vec x=0$

$\begin{bmatrix}
-1 & 1 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}=0$

$\begin{bmatrix}
-x_1+x_2 \\
0
\end{bmatrix}=0$

$-x_1+x_2=0$

$x_2=x_1$

So any vector where $x_2=x_1$ will be an eigenvector with an eigenvalue of $2$, e.g., $\begin{bmatrix}1\\1\end{bmatrix}$.

But can these eigenvectors make an eigenbasis?

* It's a 2-dimensional transform so we need at least 2 eigenvectors. CHECK!
* And they must be linearly independent. Which we can check by making sure the determinant is non-zero...

$\begin{vmatrix}
1 & 1 \\
0 & 1
\end{vmatrix}=1\times 1-1\times 0 \ne 0$ CHECK!

So, if we have this eigenbasis:

$B=\bigg\{\begin{bmatrix}1 \\0\end{bmatrix}, \begin{bmatrix}1 \\1\end{bmatrix}\bigg\}$

then we'll use the conversion matrix...

$C=
\begin{bmatrix}
1 & 1 \\
0 & 1
\end{bmatrix}$

and it's inverse:

$CC^{-1}=I$

$\begin{bmatrix}
1 & 1 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}=
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}$

$a+c=1$

$c=0$

$b+d=0$

$d=1$

$C^{-1}=\begin{bmatrix}
1 & -1 \\
0 & 1
\end{bmatrix}$

...for converting from/to this basis.

And we've already shown that the transform vector in the basis is a diagonal matrix with the relevant eigenvalues on the diagonal:

$D=\begin{bmatrix}
1 & 0 \\
0 & 2
\end{bmatrix}$

And $D^n$ is very easy to find:

$D^n=\begin{bmatrix}
1^n & 0 \\
0 & 2^n
\end{bmatrix}$

So if we want to apply the transform twice we use:

$D^2=\begin{bmatrix}
1^2 & 0 \\
0 & 2^2
\end{bmatrix}=\begin{bmatrix}
1 & 0 \\
0 & 4
\end{bmatrix}$

And finally:

$\vec v_2=CD^2C^{-1}\vec v_0$

$=\begin{bmatrix}
1 & 1 \\
0 & 1
\end{bmatrix}\begin{bmatrix}
1 & 0 \\
0 & 4
\end{bmatrix}\begin{bmatrix}
1 & -1 \\
0 & 1
\end{bmatrix}\begin{bmatrix}
-1 \\
1
\end{bmatrix}$

$=\begin{bmatrix}
1 & 4 \\
0 & 4
\end{bmatrix}\begin{bmatrix}
1 & -1 \\
0 & 1
\end{bmatrix}\begin{bmatrix}
-1 \\
1
\end{bmatrix}$

$=\begin{bmatrix}
1 & 3 \\
0 & 4
\end{bmatrix}\begin{bmatrix}
-1 \\
1
\end{bmatrix}$


$=\begin{bmatrix}
2 \\
4
\end{bmatrix}$
