The determinant of a matrix can be thought of as a description of what the matrix transform does to the area of the grids. Suppose we start off with a 2D vector space defined by unit basis vectors, $\begin{bmatrix}
1 \\
0
\end{bmatrix}$ and $\begin{bmatrix}
0 \\
1
\end{bmatrix}$. The space can be divided up into square grids with sides $1$:

![](img\determinant-before.png)

If we apply a stretching matrix transform like: 

$\begin{bmatrix}
3 & 0 \\
0 & 2
\end{bmatrix}$

...then the squares in the original space grids get stretched out to have sides $3$ and $2$:

![](img\determinant-after.png)

The **determinant** is the area of the grids: $3 \times 2=6$

If a stretching and shearing transform is applied like:

$\begin{bmatrix}
3 & 1 \\
0 & 2
\end{bmatrix}$

...then the squares in the original space end up as parallelograms:

![](img\determinant-after-shearing.png)

The area of these are $base \times height=3 \times 2=6$. The **determinant** is still 6.

For a general transform...

$\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}$

...the grid squares are transformed into quadrilaterals...

![](img\determinant-general.png)

...and it's area turns out to be: $ad-bc$

(For larger matrices it's a bit trickier to derive and we normally just rely on a computer to calculate it for us, so learning how to do it isn't too important!)

The determinant of a matrix is denoted with vertical lines:

$\begin{vmatrix}
a & b \\
c & d
\end{vmatrix}=ad-bc$

**Transform matrices that reduce the dimensionality**

Consider the transform matrix:

$\begin{bmatrix}
1 & 2 \\
1 & 2
\end{bmatrix}$

Notice how the transformed space is collapsed onto a line...

![](img\reduce-dimensionality.png)

...all $x$s will be collapsed onto the line and all $y$s will be collapsed onto the line, i.e., it's reduced the dimensionality of the space - from 2 dimensions to 1. Every point in the 2D space will map on to some point along the line.

Interestingly, it's determinant is $0$:

$\begin{vmatrix}
1 & 2 \\
1 & 2
\end{vmatrix}=1\times2-2\times1=0$

If we had a $3\times3$ matrix describing a 3D space, and if one of the basis vectors was linearly dependent on the other two then the transformed space would actually be a plane (2D). The determinant will be 0.

Consider this vector transformation...

$\begin{bmatrix}
1 & 1 & 3 \\
1 & 2 & 4 \\
2 & 3 & 7
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
17 \\
29
\end{bmatrix}$

...which, [as we know](4.%20Solving%20linear%20equations.ipynb), is the same as:

$\begin{array}{lcl}
a & + & b & + & 3c & = & 12 \\
a & + & 2b & + & 4c & = & 17 \\
2a & + & 3b & + & 7c & = & 29
\end{array}$

You'll notice that the 3rd equation doesn't actually give us any new information. It's just equal to the sum of the other two equations. In other words, despite having 3 simultaneous equations with 3 variables, we don't have enough data to solve it.

But also, if we think of the columns in the matrix as basis vectors we can see that they're not linear independent. Column 3 = 2 x Column 1 + Column 2. In other words, it collapses the 3D space onto a plane.

Watch what happens when we try to reduce it to the row echelon form...

$\begin{bmatrix}
1 & 1 & 3 \\
1 & 2 & 4 \\
2 & 3 & 7
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
17 \\
29
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
1-1 & 2-1 & 4-3 \\
2 & 3 & 7
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
17-12 \\
29
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
2 & 3 & 7
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
29
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
2-1 & 3-1 & 7-3
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
29-12
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
2-2 & 3-2 & 7-6
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
29-24
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
0 & 1 & 1
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
5
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
0-0 & 1-1 & 1-1
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
5-5
\end{bmatrix}$

$\begin{bmatrix}
1 & 1 & 3 \\
0 & 1 & 1 \\
0 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}=
\begin{bmatrix}
12 \\
5 \\
0
\end{bmatrix}$

Oh!

The matrix has no inverse (it's called singular) and its determinant is 0. Which makes sense - the inverse lets you "undo" the transformation (i.e., get from the new vectors back to the original vectors) but you can't do that if you've reduced the dimensions.
